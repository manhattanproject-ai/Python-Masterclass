# List

What is a List in Python?

A List in Python is a versatile and commonly used data structure that represents an ordered sequence of items. Think of it as a container that can hold various things, and these things stay in the order you put them in.

Key Characteristics of Python Lists:

* Ordered: The items in a list have a specific order, and this order is maintained. The position of each item is identified by an index, starting from 0 for the first item.
* Mutable (Changeable): You can add, remove, or change items in a list after it has been created. This makes them very flexible.
* Allows Duplicates: You can have multiple items with the same value in a list. For example, your shopping list might have "apples" listed twice if you need two bags.
* Heterogeneous: A single list can contain items of different data types (e.g., numbers, strings, boolean values). Your shopping list could have "milk" (string), 2 (integer), and True (boolean, maybe indicating if you have a coupon).
* Dynamic Size: Lists can grow or shrink in size as you add or remove items.

# 1. Define a Utility function to print highlighted characteristics

```python
def highlight_characteristic(name, description, symbol='='):
    """Helper function to print highlighted characteristics."""
    print(f"\n{symbol*5} {name.upper()} {symbol*5}")
    print(description)
    print(f"{symbol*5} {name.upper()} {symbol*5}\n")

print("--- Python List Data Structure: Shopping List Example ---")
```

# 2. Exploring the Ordered nature of Lists

Ordered:

* Items are added sequentially (append).
* Accessing items by their index (shopping_list[0]) always retrieves the same item.
* insert() demonstrates that items shift to accommodate new elements, maintaining the relative order of existing elements.

```python
# 1. Ordered Characteristic
highlight_characteristic(
    "Ordered",
    "Items in a list maintain their insertion order. The position of an item matters."
)

shopping_list = []
print(f"Initial shopping list: {shopping_list}")

print("Adding items in specific order:")
shopping_list.append("Milk")        # Index 0
shopping_list.append("Bread")       # Index 1
shopping_list.append("Eggs")        # Index 2
shopping_list.append("Apples")      # Index 3
print(f"Shopping list after appends: {shopping_list}")
print(f"Item at index 0: {shopping_list[0]}")
print(f"Item at index 2: {shopping_list[2]}")

# Demonstrate that order is preserved even after adding to middle
shopping_list.insert(1, "Cheese") # Inserts "Cheese" at index 1, shifting others
print(f"Shopping list after inserting 'Cheese' at index 1: {shopping_list}")
print(f"New item at index 1: {shopping_list[1]}")
print(f"Original item at index 1 (Bread) is now at index 2: {shopping_list[2]}")
```

# 3. Exploring the mutable nature of List

Mutable:

* shopping_list[0] = "Almond Milk" directly modifies an element at a specific index.
* shopping_list.remove("Eggs") deletes an item by its value.
* del shopping_list[3] deletes an item by its index.
* append() and extend() add new items, demonstrating mutability in terms of adding elements.

```python
# 2. Mutable Characteristic

highlight_characteristic(
    "Mutable",
    "List items can be changed (modified, added, removed) after the list is created."
)

print(f"Original shopping list: {shopping_list}")

# Modifying an item
shopping_list[0] = "Almond Milk"
print(f"After modifying 'Milk' to 'Almond Milk': {shopping_list}")

# Removing an item
shopping_list.remove("Eggs") # Removes the first occurrence of 'Eggs'
print(f"After removing 'Eggs': {shopping_list}")

# Removing by index
del shopping_list[3] # Removes 'Bread' (now at index 3 after previous removals/inserts)
print(f"After deleting item at index 3 (which was Bread): {shopping_list}")

# Adding new items (also demonstrates Dynamic Size)
shopping_list.append("Butter")
shopping_list.extend(["Grapes", "Potatoes"])
print(f"After adding more items: {shopping_list}")
```

# 4. Exploring Duplicates in List

Allows Duplicates:

* shopping_list.append("Almond Milk") and shopping_list.append("Apples") clearly show that the same item can exist multiple times in the list.
* shopping_list.count() is used to verify the number of occurrences.

```python
# 3. Allows Duplicates Characteristic

highlight_characteristic(
    "Allows Duplicates",
    "A list can contain multiple occurrences of the same item."
)

print(f"Current shopping list: {shopping_list}")
shopping_list.append("Almond Milk") # Adding a duplicate of an existing item
shopping_list.append("Apples")      # Adding another duplicate
print(f"Shopping list with duplicates: {shopping_list}")

print(f"Count of 'Almond Milk': {shopping_list.count('Almond Milk')}")
print(f"Count of 'Apples': {shopping_list.count('Apples')}")
```

# 5. Exploring Heterogeneous Characteristics of List

Heterogeneous:

* The shopping_list is shown containing str (strings like "Milk"), float (2.5), int (10), bool (True), list (["Soap", "Detergent"]), and even a dict ({"item": "Coffee", "weight": "250g"}).
* type() function is used to confirm the data types of different elements.

```python
# 4. Heterogeneous Characteristic

highlight_characteristic(
    "Heterogeneous",
    "A list can store items of different data types (e.g., strings, integers, floats, booleans, other lists)."
)

shopping_list.append(2.5)          # Float (e.g., price per kg)
shopping_list.append(10)           # Integer (e.g., quantity)
shopping_list.append(True)         # Boolean (e.g., discount available)
shopping_list.append(["Soap", "Detergent"]) # Another list (e.g., cleaning supplies sub-list)
shopping_list.append({"item": "Coffee", "weight": "250g"}) # Dictionary

print(f"Shopping list with heterogeneous items: {shopping_list}")
print(f"Type of item at index {len(shopping_list)-4}: {type(shopping_list[len(shopping_list)-4])}")
print(f"Type of item at index {len(shopping_list)-1}: {type(shopping_list[len(shopping_list)-1])}")
```

# 6. Exploring the Dynamic size nature of List

Dynamic Size:

* The dynamic_shopping_list starts empty.
* append() increases its size.
* pop() and remove() decrease its size.
* extend() dramatically increases its size in one go.
* len() function is used at each step to demonstrate the changing size without any pre-defined limit.
* clear() resets the size to zero.

```python
# 5. Dynamic Size Characteristic

highlight_characteristic(
    "Dynamic Size",
    "A list can grow or shrink in size as elements are added or removed, without needing to pre-define its capacity."
)

dynamic_shopping_list = []
print(f"Initial dynamic list (size {len(dynamic_shopping_list)}): {dynamic_shopping_list}")

print("Adding 3 items...")
dynamic_shopping_list.append("Shampoo")
dynamic_shopping_list.append("Conditioner")
dynamic_shopping_list.append("Toothpaste")
print(f"After adding items (size {len(dynamic_shopping_list)}): {dynamic_shopping_list}")

print("Removing 1 item...")
dynamic_shopping_list.pop() # Removes last item
print(f"After removing item (size {len(dynamic_shopping_list)}): {dynamic_shopping_list}")

print("Adding many items with extend...")
more_items = ["Toilet Paper", "Soap", "Deodorant", "Paper Towels", "Dish Soap"]
dynamic_shopping_list.extend(more_items)
print(f"After extending (size {len(dynamic_shopping_list)}): {dynamic_shopping_list}")

# Clear the list
dynamic_shopping_list.clear()
print(f"After clearing the list (size {len(dynamic_shopping_list)}): {dynamic_shopping_list}")
```
