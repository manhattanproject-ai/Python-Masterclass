# Dictionary

What is a Dictionary in Python?

A Dictionary in Python is an unordered collection of key-value pairs. Each key in a dictionary is unique, and it's used to access its corresponding value.


Key Characteristics of Python Dictionaries:

* Unordered: Unlike lists and tuples, dictionaries do not maintain a specific order of items. The order in which you add key-value pairs is not necessarily the order in which they are stored or retrieved (though in Python 3.7+, dictionaries remember insertion order as an implementation detail, but you shouldn't rely on it for logical ordering).
* Key-Value Pairs: Each item in a dictionary consists of a unique key and an associated value.
* Unique Keys: Keys within a dictionary must be unique. If you try to add a new value with an existing key, the old value associated with that key will be overwritten.
* Mutable (Changeable): You can add, remove, and modify key-value pairs in a dictionary after it has been created.
* Heterogeneous Values: The values in a dictionary can be of any data type (numbers, strings, lists, other dictionaries, etc.).
* Keys Must Be Hashable: Keys must be of an immutable and hashable type, such as strings, numbers, or tuples (if they only contain immutable elements). Lists cannot be used as keys because they are mutable.

# 1. Define a Utility function to print highlighted characteristics

```python
import collections

def highlight_characteristic(name, description, symbol='='):
    """Helper function to print highlighted characteristics."""
    print(f"\n{symbol*5} {name.upper()} {symbol*5}")
    print(description)
    print(f"{symbol*5} {name.upper()} {symbol*5}\n")

print("--- Python Dictionary Data Structure: Product Catalog Example ---")
```

# 2. Explain Key-Value Pairs Characteristic of Dictionaries

Key-Value Pairs:

* The product_catalog clearly demonstrates this: "P001" is a key, and {"name": "Laptop Pro", ...} is its associated value.
* Elements are accessed using the key within square brackets (e.g., product_catalog['P001']).
* Attempting to access a non-existent key raises a KeyError.

```python
# 1. Key-Value Pairs Characteristic

highlight_characteristic(
    "Key-Value Pairs",
    "Dictionaries store data as unique 'keys' mapped to corresponding 'values'."
)

# Creating a product catalog where product_ID is the key and product details are the value
product_catalog = {
    "P001": {"name": "Laptop Pro", "price": 1200.00, "stock": 50},
    "P002": {"name": "Wireless Mouse", "price": 25.50, "stock": 200},
    "P003": {"name": "External SSD 1TB", "price": 99.99, "stock": 120}
}
print(f"Initial Product Catalog: {product_catalog}")

print("\nAccessing values using keys:")
print(f"Details for P001: {product_catalog['P001']}")
print(f"Price of Wireless Mouse (P002): {product_catalog['P002']['price']}")

# Attempting to access a non-existent key (will cause an error)
try:
    print(f"Details for P004: {product_catalog['P004']}")
except KeyError as e:
    print(f"Error caught as expected: {e}")
    print("Accessing a non-existent key raises a KeyError.\n")
```

# 3. Explain Mutable (Changeable) Characteristic of Dictionaries

Mutable (Changeable):

* Adding: product_catalog["P004"] = ... demonstrates adding a new key-value pair.
* Updating: product_catalog["P001"]["price"] = ... and product_catalog["P003"]["stock"] -= 10 show how values associated with existing keys can be modified.
* Removing: del product_catalog["P002"] removes a key-value pair by key, and product_catalog.pop("P004") removes a pair while returning its value.

```python
# 2. Mutable (Changeable) Characteristic

highlight_characteristic(
    "Mutable (Changeable)",
    "Dictionaries can be modified after creation: new key-value pairs can be added, values can be updated, and pairs can be removed."
)

print(f"Current Product Catalog: {product_catalog}")

# Adding a new product
product_catalog["P004"] = {"name": "Mechanical Keyboard", "price": 75.00, "stock": 80}
print(f"After adding new product 'P004': {product_catalog}")

# Updating an existing product's value
product_catalog["P001"]["price"] = 1150.00 # Price drop for Laptop Pro
product_catalog["P003"]["stock"] -= 10   # 10 SSDs sold
print(f"After updating price of 'P001' and stock of 'P003': {product_catalog}")

# Removing a product
del product_catalog["P002"] # Wireless Mouse discontinued
print(f"After removing 'P002': {product_catalog}")

# Using pop() to remove and get the value
removed_product = product_catalog.pop("P004")
print(f"After removing 'P004' using pop(): {product_catalog}")
print(f"Removed product details: {removed_product}\n")
```

# 4. Explain ordered insertion

Ordered (Insertion Order - Modern Python):

* The ordered_catalog example explicitly shows items being added in a specific order.
* When iterated over (using a for loop), the keys (and thus their values) are retrieved in the exact order they were inserted.
* Even after adding and deleting items, the relative order of the remaining items is preserved, and newly added items appear at the end. This is a crucial distinction from traditional "unordered" data structures.

```python
# 3. Ordered (Insertion Order) Characteristic (Modern Python)

highlight_characteristic(
    "Ordered (Insertion Order - Modern Python)",
    "As of Python 3.7+, dictionaries maintain the order in which key-value pairs are inserted. This means when you iterate, they appear in the order they were added."
)

ordered_catalog = {}
ordered_catalog["Book"] = 25.99
ordered_catalog["Pen"] = 1.50
ordered_catalog["Notebook"] = 10.00
ordered_catalog["Eraser"] = 0.75

print(f"Catalog added in specific order: {ordered_catalog}")
print("\nIterating through keys to show insertion order:")
for key in ordered_catalog:
    print(key, end=" -> ")
print("...\n")

# If you add or delete items, the insertion order for the remaining items is preserved,
# and new items are added at the end.
ordered_catalog["Pencil"] = 0.50
del ordered_catalog["Book"]
print(f"Catalog after adding 'Pencil' and deleting 'Book': {ordered_catalog}")
print("\nIterating again:")
for key in ordered_catalog:
    print(key, end=" -> ")
print("...\n")
```

# 5. Explain Heterogeneous nature of Dictionaries

Heterogeneous Values:

* The product_info dictionary contains values of various Python data types: str, bool, float, int, list, dict, and NoneType. This proves that dictionary values are not restricted to a single type.

```python
# 4. Heterogeneous Values Characteristic

highlight_characteristic(
    "Heterogeneous Values",
    "The values in a dictionary can be of different data types."
)

product_info = {
    "product_id": "PNL987",           # String
    "is_available": True,             # Boolean
    "current_price": 75.25,           # Float
    "reviews_count": 150,             # Integer
    "features": ["Waterproof", "Durable", "Lightweight"], # List
    "supplier_details": {             # Dictionary
        "name": "Tech Corp",
        "contact": "supplier@example.com"
    },
    "launch_date": None               # NoneType
}
print(f"Product info with heterogeneous values: {product_info}")
print(f"Type of 'is_available': {type(product_info['is_available'])}")
print(f"Type of 'features': {type(product_info['features'])}")
print(f"Type of 'supplier_details': {type(product_info['supplier_details'])}\n")
```

# 6. Keys must be hashable in dictionaries

* The valid_keys_dict demonstrates that immutable types like strings, integers, floats, booleans, and tuples (containing only immutable elements) can serve as keys.
* The try-except blocks explicitly show that using mutable types like list or dict as keys results in a TypeError, because they are unhashable. Hashability is a requirement for dictionary keys to allow for efficient lookup.

```python
# 5. Keys Must Be Hashable Characteristic

highlight_characteristic(
    "Keys Must Be Hashable",
    "Dictionary keys must be immutable (unchangeable) types like strings, numbers, or tuples (if the tuple contains only immutable elements). Mutable types like lists or dictionaries cannot be used as keys."
)

# Examples of valid (hashable) keys:
valid_keys_dict = {
    "product_name": "Chair",        # String (Hashable)
    101: "Customer ID",             # Integer (Hashable)
    3.14: "Pi Value",               # Float (Hashable)
    True: "Boolean Key",            # Boolean (Hashable)
    (1, "sku_xyz"): "Product SKU Tuple", # Tuple with immutable elements (Hashable)
    None: "Null Value Key"          # NoneType (Hashable)
}
print(f"Dictionary with valid (hashable) keys: {valid_keys_dict}")

# Examples of invalid (unhashable) keys (will cause TypeError):
print("\nAttempting to use unhashable types as keys (will cause TypeError):")
try:
    invalid_keys_dict = {
        ["invalid_list_key"]: "This won't work" # List is mutable, hence unhashable
    }
except TypeError as e:
    print(f"Error caught for list as key: {e}")

try:
    invalid_keys_dict = {
        {"invalid_dict_key": "data"}: "Nor will this" # Dictionary is mutable, hence unhashable
    }
except TypeError as e:
    print(f"Error caught for dictionary as key: {e}")
```
