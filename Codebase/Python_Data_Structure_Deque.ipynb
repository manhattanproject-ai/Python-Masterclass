# Deque

What is deque in Python?

deque (pronounced "deck") stands for double-ended queue. It's a list-like data structure provided by Python's collections module that allows you to efficiently add and remove elements from both ends (left and right).
Think of it as a more general version of a queue and a stack combined, but optimized for operations at its ends.

Key Characteristics of deque:

* Ordered: Elements in a deque maintain the order in which they were added.
* Mutable: You can add, remove, and extend elements in a deque.
* Allows Duplicates: A deque can contain multiple identical elements.
* Heterogeneous: It can hold elements of different data types.
* Efficient Appends and Pops from Both Ends: This is the key advantage over regular Python lists. Appending and popping from either end of a deque takes O(1) constant time, making it very fast even for large collections. For regular lists, appending to the end is O(1), but inserting or removing from the beginning is O(n) (where n is the number of elements) because all other elements need to be shifted.

# 1. Define a Utility function to print highlighted characteristics

```python
import collections

def highlight_characteristic(name, description, symbol='='):
    """Helper function to print highlighted characteristics."""
    print(f"\n{symbol*5} {name.upper()} {symbol*5}")
    print(description)
    print(f"{symbol*5} {name.upper()} {symbol*5}\n")

print("--- Python Deque Data Structure: Task Management System Example ---")
```

```python
# Initialize a deque for managing tasks
# Each task will be a string for simplicity, but could be more complex objects

task_queue = collections.deque()
print(f"Initial Task Queue: {list(task_queue)}") # Convert to list for clear printing
```

# 2. Ordered characteristics of deque

Ordered:

* Elements are added and retrieved in a specific sequence, whether using append() (to the right/end) or appendleft() (to the left/beginning).
* Iterating through the deque (e.g., for task in task_queue:) always yields elements in their current, established order.

```python
# 1. Ordered Characteristic

highlight_characteristic(
    "Ordered",
    "Elements in a deque maintain their insertion order. When you add elements, they appear in a predictable sequence, and iteration follows this sequence."
)

print("Adding tasks to the right (end of the queue):")
task_queue.append("Task A: Prepare Presentation")
task_queue.append("Task B: Send Meeting Invites")
task_queue.append("Task C: Review Project Plan")
print(f"Tasks after appending: {list(task_queue)}")

print("\nAdding an urgent task to the left (front of the queue):")
task_queue.appendleft("Task Urgent: Fix Critical Bug")
print(f"Tasks after appending left: {list(task_queue)}")

print("\nIterating through tasks (order is preserved):")
for i, task in enumerate(task_queue):
    print(f"  {i+1}. {task}")
```

# 3. Mutable nature of deque

Mutable (Changeable):

* task_queue.append() and task_queue.appendleft() demonstrate adding elements.
* task_queue.pop() and task_queue.popleft() demonstrate removing elements from either end.
* task_queue[1] = "..." shows that individual elements can be modified by index (though direct arbitrary indexing for access or slicing is not the primary use case and less efficient than end operations).
* task_queue.extend() adds multiple elements.
* task_queue.clear() empties the deque.

```python
# 2. Mutable (Changeable) Characteristic

highlight_characteristic(
    "Mutable (Changeable)",
    "A deque can be modified after creation: elements can be added, removed, or their values can be updated."
)

print(f"Current Task Queue: {list(task_queue)}")

# Modifying a task (deques support indexing for modification, though not for direct access)
# Note: Direct indexing for modification is supported, but not for arbitrary access/slicing
task_queue[1] = "Task B (Updated): Confirm Meeting Agenda"
print(f"After modifying 'Task B': {list(task_queue)}")

# Removing tasks from both ends (also demonstrates efficiency)
completed_urgent_task = task_queue.popleft()
print(f"Removed (completed) urgent task from left: '{completed_urgent_task}'")
print(f"Queue after popleft: {list(task_queue)}")

completed_regular_task = task_queue.pop()
print(f"Removed (completed) regular task from right: '{completed_regular_task}'")
print(f"Queue after pop: {list(task_queue)}")

# Adding more elements (also demonstrates dynamic size)
task_queue.extend(["Task D: Follow up with Client", "Task E: Draft Report"])
print(f"After extending with new tasks: {list(task_queue)}")

# Clear the deque
task_queue.clear()
print(f"After clearing the queue: {list(task_queue)} (Empty deque)")
```

# 4. Allows duplicates in deque

Allows Duplicates:

* The task_queue is shown explicitly containing "Task F: Data Entry" and "Task G: Code Review" multiple times, demonstrating that deques do not enforce uniqueness.
* task_queue.count() is used to verify the number of occurrences.

```python
# 3. Allows Duplicates Characteristic

highlight_characteristic(
    "Allows Duplicates",
    "A deque can contain multiple identical elements."
)

task_queue.append("Task F: Data Entry")
task_queue.append("Task G: Code Review")
task_queue.append("Task F: Data Entry") # Duplicate task
task_queue.appendleft("Task G: Code Review") # Another duplicate task
print(f"Task Queue with duplicate tasks: {list(task_queue)}")
print(f"Count of 'Task F: Data Entry': {task_queue.count('Task F: Data Entry')}")
print(f"Count of 'Task G: Code Review': {task_queue.count('Task G: Code Review')}")
```

# 5. deque are heterogeneous in nature

Heterogeneous Values:

* The mixed_tasks_and_data deque contains a str, int, bool, float, dict, and tuple, illustrating that it can store elements of different data types.

```python
# 4. Heterogeneous Values Characteristic

highlight_characteristic(
    "Heterogeneous Values",
    "A deque can store elements of different data types."
)

mixed_tasks_and_data = collections.deque()
mixed_tasks_and_data.append("Task H: Brainstorm Ideas") # String
mixed_tasks_and_data.append(101)                       # Integer (e.g., project ID)
mixed_tasks_and_data.append(True)                      # Boolean (e.g., high priority)
mixed_tasks_and_data.append(3.14)                      # Float (e.g., completion percentage)
mixed_tasks_and_data.append({"assignee": "Alice", "due_date": "2025-06-15"}) # Dictionary
mixed_tasks_and_data.append(("Review", "Finalize"))    # Tuple
print(f"Deque with heterogeneous values: {list(mixed_tasks_and_data)}")
print(f"Type of element at index 1: {type(mixed_tasks_and_data[1])}")
print(f"Type of element at index 4: {type(mixed_tasks_and_data[4])}")
```

# 6. Efficient Appends and Pops from Both Ends Characteristic

Efficient Appends and Pops from Both Ends:

* This is the core strength of deque. The example simulates adding and removing a large number of tasks (1,000,000) from both ends. These operations are performed in O(1) time complexity, meaning their speed does not significantly degrade as the deque grows larger. This is a key advantage over Python lists, where insert(0, item) or pop(0) would be O(n) (linear time), becoming very slow for large lists.
* The rotate() method is also highlighted as a unique and efficient operation specific to deques for shifting elements.

```python
# 5. Efficient Appends and Pops from Both Ends Characteristic

highlight_characteristic(
    "Efficient Appends and Pops from Both Ends",
    "This is the primary advantage of deques. Adding or removing elements from either end is highly efficient (O(1) time complexity), unlike lists where operations at the beginning can be slow (O(n))."
)

large_task_queue = collections.deque()

# Simulate adding many tasks
print("Simulating adding 1,000,000 tasks...")
for i in range(500000):
    large_task_queue.append(f"RegTask-{i}")
for i in range(500000):
    large_task_queue.appendleft(f"UrgentTask-{i}")
print(f"Large queue created. Size: {len(large_task_queue)}")

# Simulate processing tasks from both ends
print("Processing 10 tasks from the right (pop):")
for _ in range(10):
    large_task_queue.pop()
print(f"Queue size after 10 pops: {len(large_task_queue)}")

print("Processing 10 tasks from the left (popleft):")
for _ in range(10):
    large_task_queue.popleft()
print(f"Queue size after 10 poplefts: {len(large_task_queue)}")

print("These operations (appendleft/popleft/append/pop) are very fast, regardless of deque size.")

# Deques also support rotation, which is efficient:
highlight_characteristic(
    "Rotation (Deque Specific)",
    "Deques can efficiently rotate elements from one end to the other.", symbol='-')

rotation_queue = collections.deque(["A", "B", "C", "D", "E"])
print(f"Original queue for rotation: {list(rotation_queue)}")
rotation_queue.rotate(1) # Shifts elements to the right by 1 (E moves to front)
print(f"After rotate(1): {list(rotation_queue)}")
rotation_queue.rotate(-2) # Shifts elements to the left by 2 (C, D move to back)
print(f"After rotate(-2): {list(rotation_queue)}")
```
