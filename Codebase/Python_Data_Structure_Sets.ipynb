# Sets

What is a Set in Python?

A Set in Python is an unordered collection of unique elements. Think of it as a bag where each item inside is distinct, and the way the items are arranged in the bag doesn't have any specific meaning.

Key Characteristics of Python Sets:

* Unordered: The items in a set do not have a specific order. You cannot access elements by index like you do in lists or tuples. The order can even change each time you print the same set.
* Unique Elements: Sets automatically remove duplicate items. If you try to add an element that already exists in the set, it won't be added again.
* Mutable (Changeable): You can add or remove elements from a set after it has been created.
* Heterogeneous: A set can contain items of different data types (as long as they are hashable - immutable types like numbers, strings, and tuples).
* Supports Set Operations: Sets in Python provide efficient ways to perform common mathematical set operations like union, intersection, difference, and symmetric difference.

# 1. Define a Utility function to print highlighted characteristics

```python
import collections

def highlight_characteristic(name, description, symbol='='):
    """Helper function to print highlighted characteristics."""
    print(f"\n{symbol*5} {name.upper()} {symbol*5}")
    print(description)
    print(f"{symbol*5} {name.upper()} {symbol*5}\n")

print("--- Python Set Data Structure: Project Skills Example ---")
```

# 2. Exploring the UnOrdered nature of Sets

Unordered:

* project_skills_alpha and project_skills_beta are created with the same elements but in different literal order. They are == True when compared, demonstrating that order doesn't matter for set equality.
* An attempt to access an element by index (project_skills_alpha[0]) results in a TypeError, proving that sets do not support indexing.

```python
# 1. Unordered Characteristic

highlight_characteristic(
    "Unordered",
    "Elements in a set do not have a defined order. You cannot access elements by index or slice them."
)

project_skills_alpha = {"Python", "SQL", "Cloud Computing", "Machine Learning", "Data Visualization"}
project_skills_beta = {"Cloud Computing", "Data Visualization", "SQL", "Python", "Machine Learning"}

print(f"Project Alpha Skills: {project_skills_alpha}")
print(f"Project Beta Skills (same elements, different literal order): {project_skills_beta}")
print(f"Are they considered equal? {project_skills_alpha == project_skills_beta}") # True, because order doesn't matter for equality

# Attempting to access by index (will cause an error)
try:
    print(f"Attempting to access element at index 0: {project_skills_alpha[0]}")
except TypeError as e:
    print(f"Error caught as expected: {e}")
    print("This confirms sets are unordered and do not support indexing.\n")
```

# 3. Exploring the Unique Elements nature of Sets

Unique Elements:

* When team_member_skills is initialized with {"Python", "SQL", "Python", "Java", "SQL", "Agile"}, the output {'Agile', 'Java', 'Python', 'SQL'} clearly shows that Python and SQL appear only once, as duplicates are automatically discarded.
* Attempting to add() an element already present in the set (e.g., "SQL") has no effect on the set's contents or size, further demonstrating uniqueness.

```python
# 2. Unique Elements Characteristic

highlight_characteristic(
    "Unique Elements",
    "A set automatically ensures that all its elements are unique. Duplicate elements are automatically removed."
)

# Trying to create a set with duplicates
team_member_skills = {"Python", "SQL", "Python", "Java", "SQL", "Agile"}
print(f"Skills provided (with duplicates): Python, SQL, Python, Java, SQL, Agile")
print(f"Set of skills created from input: {team_member_skills}")
print(f"Note how 'Python' and 'SQL' only appear once, confirming uniqueness.")

# Adding a duplicate element to an existing set
print(f"Current unique skills: {team_member_skills}")
team_member_skills.add("SQL") # Adding an element already present
print(f"After attempting to add 'SQL' again: {team_member_skills} (No change)")
print(f"Size of set: {len(team_member_skills)}") # Size remains the same
```

# 4. Exploring Mutable nature  in Sets

Mutable:

* team_member_skills.add("Docker") demonstrates adding a new element.
* team_member_skills.remove("Java") and team_member_skills.discard("Agile") show how elements can be removed from the set.
* team_member_skills.clear() removes all elements, leaving an empty set. This confirms that the set itself can be modified after creation.

```python
# 3. Mutable Characteristic

highlight_characteristic(
    "Mutable",
    "Sets can be changed after creation: elements can be added or removed. (However, individual elements cannot be modified in-place)."
)

print(f"Current team member skills: {team_member_skills}")

# Adding a new skill
team_member_skills.add("Docker")
print(f"After adding 'Docker': {team_member_skills}")

# Removing a skill
team_member_skills.remove("Java") # Using remove()
print(f"After removing 'Java': {team_member_skills}")

team_member_skills.discard("Agile") # Using discard()
print(f"After discarding 'Agile': {team_member_skills}")

# Clear the set
team_member_skills.clear()
print(f"After clearing the set: {team_member_skills} (Empty set)")
```

# 5. Exploring Heterogeneous Characteristics of Sets

Heterogeneous:

* The mixed_requirements set contains elements of different data types: str, int, float, bool, and even an immutable tuple.
* A comment highlights that mutable types like list or dict cannot be directly stored as elements in a set because set elements must be "hashable."

```python
# 4. Heterogeneous Characteristic

highlight_characteristic(
    "Heterogeneous",
    "A set can store elements of different, hashable data types."
)

# A set representing mixed project requirements/metadata
mixed_requirements = {
    "Backend Development", # String
    2025,                  # Integer (e.g., target year)
    3.14,                  # Float (e.g., version number)
    True,                  # Boolean (e.g., is critical task?)
    ("API Design", "Security"), # Tuple (immutable, hence hashable)
    # Note: Lists and dictionaries are mutable, so they cannot be elements of a set directly.
    # {"framework": "React"} # This would cause a TypeError
}
print(f"Mixed requirements set: {mixed_requirements}")
print(f"Type of element '2025': {type(2025)}")
print(f"Type of element ('API Design', 'Security'): {type(('API Design', 'Security'))}\n")
```

# 6. Exploring Set operations in Sets

Supports Set Operations:

* The example showcases several common set operations using both method calls (.union(), .intersection(), .difference(), .symmetric_difference(), .issubset(), .issuperset()) and their corresponding operator shortcuts (|, &, -, ^).
* These operations demonstrate how sets are powerful for comparing and combining collections of unique items, which is very useful in scenarios like comparing skill sets, finding commonalities, or identifying gaps.

```python
# 5. Supports Set Operations Characteristic

highlight_characteristic(
    "Supports Set Operations",
    "Sets provide powerful methods for mathematical set operations like union, intersection, difference, and symmetric difference."
)

team_alpha_skills = {"Python", "SQL", "Cloud Computing", "Leadership", "Project Management"}
team_beta_skills = {"Java", "SQL", "Machine Learning", "Cloud Computing", "DevOps"}
project_needs = {"Python", "Machine Learning", "Data Visualization", "Cloud Computing", "Agile"}

print(f"Team Alpha Skills: {team_alpha_skills}")
print(f"Team Beta Skills: {team_beta_skills}")
print(f"Project Needs: {project_needs}\n")

# Union: All skills from both teams
all_team_skills = team_alpha_skills.union(team_beta_skills)
print(f"Union (All skills from both teams): {all_team_skills}")
print(f"Using '|' operator: {team_alpha_skills | team_beta_skills}\n")

# Intersection: Common skills between the two teams
common_skills = team_alpha_skills.intersection(team_beta_skills)
print(f"Intersection (Common skills between teams): {common_skills}")
print(f"Using '&' operator: {team_alpha_skills & team_beta_skills}\n")

# Difference: Skills Team Alpha has that Team Beta does NOT have
alpha_only_skills = team_alpha_skills.difference(team_beta_skills)
print(f"Difference (Skills Alpha has, Beta does not): {alpha_only_skills}")
print(f"Using '-' operator: {team_alpha_skills - team_beta_skills}\n")

# Symmetric Difference: Skills that are in either set, but not in their intersection (unique to each)
unique_to_each_team = team_alpha_skills.symmetric_difference(team_beta_skills)
print(f"Symmetric Difference (Skills unique to each team): {unique_to_each_team}")
print(f"Using '^' operator: {team_alpha_skills ^ team_beta_skills}\n")

# Subset: Is Team Alpha's skills a subset of Project Needs?
is_subset = team_alpha_skills.issubset(project_needs)
print(f"Is Team Alpha's skills a subset of Project Needs? {is_subset}")
print(f"('Leadership' and 'Project Management' in Alpha are not in Project Needs)\n")

# Superset: Does Project Needs include all of Team Beta's skills?
is_superset = project_needs.issuperset(team_beta_skills)
print(f"Does Project Needs include all of Team Beta's skills? {is_superset}")
print(f"('Java' and 'DevOps' in Beta are not in Project Needs)\n")

# Finding missing skills for a project:
missing_project_skills_from_alpha = project_needs.difference(team_alpha_skills)
print(f"Project skills missing from Team Alpha: {missing_project_skills_from_alpha}")
```
