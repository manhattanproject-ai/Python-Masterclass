# 1. Control statement - Branching / Jump Statements (break: Terminates the current loop entirely and transfers control to the statement immediately following the loop.)

* inventory = [...]: We define a list representing our store's product inventory.
* target_product = "Monitor": This is the item we want to find.
* product_found = False: A boolean variable (a "flag") is initialized to keep track of whether the product was found.
* for product in inventory:: This for loop starts iterating through each product in the inventory list.
* if product == target_product:: Inside the loop, in each iteration, this condition checks if the current product matches our target_product.
  * When product becomes "Monitor": The condition product == target_product becomes True.
* break: Because the condition is True when product is "Monitor", the code inside the if block executes. As soon as break is encountered, the for loop is immediately terminated.
  * Notice that the print(f"Checking: ...") for "Webcam", "Headphones", etc., will not be executed, even though there are more items in the inventory list. The loop stops dead in its tracks.
* if product_found:: After the loop finishes (either by break or by iterating through all items), this if-else statement determines the final message based on the product_found flag.


```python
# --- Example: Searching for a specific product in a store inventory ---

# Our inventory (a list of products)
inventory = [
    "Laptop",
    "Mouse",
    "Keyboard",
    "Monitor",
    "Webcam",
    "Headphones",
    "Printer",
    "Speakers"
]

# The product we are looking for
target_product = "Monitor"
# target_product = "Hard Drive" # Uncomment this line to test a product not in the list

print(f"--- Searching for '{target_product}' in the inventory ---")

product_found = False # Flag to track if the product was found

# Use a 'for' loop to go through each product in the inventory
for product in inventory:
    print(f"Checking: {product}") # This shows the current product being checked

    # Condition: Check if the current product is the one we are looking for
    if product == target_product:
        print(f"\n--- Found '{target_product}'! ---")
        product_found = True # Set the flag to True
        break # <-- This is the 'break' statement!
              # It immediately exits the 'for' loop.
              # The loop will not check 'Webcam', 'Headphones', etc.

# This line of code is outside and after the 'for' loop.
# It will always execute once the loop has finished or was broken.
if product_found:
    print("Search complete. Product was located.")
else:
    print("Search complete. Product was not found in the inventory.")

print("\n--- End of program ---")
```

How it works with different target_product values:

* If target_product = "Monitor": The loop will run for "Laptop", "Mouse", "Keyboard", "Monitor". When "Monitor" is found, break activates, and the loop stops.
* If target_product = "Hard Drive" (uncomment that line): The loop will run for all items in the inventory. Since "Hard Drive" is never found, the if product == target_product: condition will never be True, so break is never executed. The product_found flag remains False, and the final else block will execute.

```python
# --- Example: Searching for a specific product in a store inventory ---

# Our inventory (a list of products)
inventory = [
    "Laptop",
    "Mouse",
    "Keyboard",
    "Monitor",
    "Webcam",
    "Headphones",
    "Printer",
    "Speakers"
]

# The product we are looking for
# target_product = "Monitor"
target_product = "Hard Drive" # Uncomment this line to test a product not in the list

print(f"--- Searching for '{target_product}' in the inventory ---")

product_found = False # Flag to track if the product was found

# Use a 'for' loop to go through each product in the inventory
for product in inventory:
    print(f"Checking: {product}") # This shows the current product being checked

    # Condition: Check if the current product is the one we are looking for
    if product == target_product:
        print(f"\n--- Found '{target_product}'! ---")
        product_found = True # Set the flag to True
        break # <-- This is the 'break' statement!
              # It immediately exits the 'for' loop.
              # The loop will not check 'Webcam', 'Headphones', etc.

# This line of code is outside and after the 'for' loop.
# It will always execute once the loop has finished or was broken.
if product_found:
    print("Search complete. Product was located.")
else:
    print("Search complete. Product was not found in the inventory.")

print("\n--- End of program ---")
```

# 2. Control statement - Branching / Jump Statements (continue: Skips the rest of the current iteration of the loop and proceeds to the next iteration.)

* numbers = [...]: We define a list containing both positive, negative, and zero numbers.
* for num in numbers:: This for loop iterates through each num in the numbers list.
* print(f"\nChecking number: {num}"): This line executes at the beginning of every iteration, showing which number is currently being considered.
* if num <= 0:: This is the condition that determines whether to skip the rest of the current iteration.
  * When num is -5 or 0 or -2: This condition becomes True.
* continue: When if num <= 0: is True, the print statement indicating the skip is executed, and then continue is encountered.
  * As soon as continue runs, Python immediately jumps back to the top of the for loop. It does not execute square = num * num or print(f" {num} is positive...") for that particular iteration. Instead, it directly fetches the next number from the numbers list (7 after 0, for example) and starts a new iteration.
* square = num * num and subsequent print: These lines are only executed if the if num <= 0: condition was False (i.e., if num was a positive number).

```python
# --- Example: Processing only positive numbers in a list ---

numbers = [10, -5, 0, 7, -2, 15, 3]

print("--- Processing numbers (skipping non-positive ones) ---")

# Use a 'for' loop to iterate through each number in the list
for num in numbers:
    # 1. Print the number being checked
    print(f"\nChecking number: {num}")

    # 2. Condition: If the number is not positive, skip the rest of this iteration
    if num <= 0:
        print(f"  {num} is not a positive number. Skipping this one.")
        continue # <-- This is the 'continue' statement!
                  # It immediately jumps to the next iteration of the loop.
                  # The code below 'continue' will NOT execute for this 'num'.

    # 3. This block will ONLY execute if 'continue' was NOT triggered
    square = num * num
    print(f"  {num} is positive. Its square is: {square}")

print("\n--- Finished processing numbers ---")
```
