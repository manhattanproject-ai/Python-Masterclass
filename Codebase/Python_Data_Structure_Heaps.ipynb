# Heaps

What is a Heap?

A Heap is a specialized tree-based data structure that satisfies the heap property. There are two main types of heaps:

* Min-Heap: For every node i, the value of the node is less than or equal to the value of its children. The smallest element is always at the root.   
* Max-Heap: For every node i, the value of the node is greater than or equal to the value of its children. The largest element is always at the root.
Heaps are typically implemented using binary trees, specifically complete binary trees (all levels are completely filled except possibly the last level, which is filled from left to right). This structure allows for efficient insertion and extraction of the minimum (in a min-heap) or maximum (in a max-heap) element.

# 1. Heap Representation

* In Python, a heap is implemented as a regular list. The heapq module provides functions (heappush, heappop, etc.) that operate on this list to maintain the heap property.
* We store tasks as (priority, task_description) tuples. heapq uses the first element of the tuple (the priority number) for comparison, ensuring the highest-priority tasks (lowest numbers) are always prioritized.
* Let's use a simple scenario: A Task Prioritization System. Imagine you have a list of tasks, and you always want to work on the one with the highest priority first. We'll assign priority numbers where lower numbers mean higher priority (e.g., priority 1 is more urgent than priority 5).

```python
import heapq # Python's built-in heap queue module

print("--- Heap Data Structure: Task Prioritization Example ---")

# 1. Initialize an empty list that will function as our min-heap
# In Python, a heap is typically a regular list that adheres to the heap property
# enforced by the `heapq` module's functions.
# We'll store tasks as tuples: (priority, task_description)
# `heapq` compares elements based on the first item in the tuple by default.
priority_queue = []

print("Initial Priority Queue (Empty):", priority_queue)
```

# 2. Adding Tasks (heappush operation)

heapq.heappush(heap, item) (Insertion):

* When you heappush an item (like (3, "Reply to urgent emails")), it's added to the list, and then the heap property is restored by "bubbling up" the new item if its priority is higher than its parent's.
* You'll notice that the list priority_queue isn't fully sorted, but the element at priority_queue[0] (the root of the conceptual tree) is always the smallest (highest priority).

```python
# 2. Adding Tasks (heappush operation)

print("\n--- Adding Tasks to the Priority Queue ---")

# Syntax: heapq.heappush(heap, item)
# Items are added in a way that maintains the heap property.
heapq.heappush(priority_queue, (3, "Reply to urgent emails"))
print(f"After adding (3, 'Reply to urgent emails'): {priority_queue}")

heapq.heappush(priority_queue, (1, "Fix critical production bug"))
print(f"After adding (1, 'Fix critical production bug'): {priority_queue}")

heapq.heappush(priority_queue, (5, "Write documentation for new feature"))
print(f"After adding (5, 'Write documentation for new feature'): {priority_queue}")

heapq.heappush(priority_queue, (2, "Prepare for team meeting"))
print(f"After adding (2, 'Prepare for team meeting'): {priority_queue}")

heapq.heappush(priority_queue, (4, "Review colleague's code"))
print(f"After adding (4, 'Review colleague's code'): {priority_queue}")

print("\n--- Current Priority Queue (List Representation of Heap) ---")
print(priority_queue)
print("Notice how the smallest priority (1) is at the front (index 0).")
print("The rest of the list follows the heap property, though it's not strictly sorted.")
```

# 3. Processing Tasks (heappop operation)

heapq.heappop(heap) (Extraction):

* This function efficiently removes and returns the smallest item from the heap (which is always at the root, priority_queue[0]).
* After removing the root, the last element in the list is moved to the root's position, and then "bubbled down" (swapped with its smaller child) until the heap property is restored.

```python
# 3. Processing Tasks (heappop operation)

print("\n--- Processing Tasks (Extracting Highest Priority) ---")

# Syntax: heapq.heappop(heap)
# This removes and returns the smallest item (highest priority) from the heap.
# The heap property is maintained afterwards.

next_task = heapq.heappop(priority_queue)
print(f"Processed task (highest priority): {next_task}")
print(f"Priority Queue after pop: {priority_queue}")

next_task = heapq.heappop(priority_queue)
print(f"Processed task (next highest priority): {next_task}")
print(f"Priority Queue after pop: {priority_queue}")

# Add another task during processing
print("\n--- Adding a new task while processing ---")
heapq.heappush(priority_queue, (0, "EMERGENCY: Server crashed!"))
print(f"Priority Queue after adding emergency task: {priority_queue}")

next_task = heapq.heappop(priority_queue)
print(f"Processed task (now the highest priority): {next_task}")
print(f"Priority Queue after pop: {priority_queue}")

print("\n--- Remaining Tasks ---")
# You can view the smallest element without popping using priority_queue[0]
if priority_queue:
    print(f"Next task to process (without removing): {priority_queue[0]}")
    print(f"All remaining tasks: {priority_queue}")
else:
    print("No tasks remaining in the queue.")
```

# Key points

Why is this efficient?

* Fastest Access to Min/Max: Finding the smallest (or largest) item is O(1) because it's always at the root.
* Efficient Insertion/Deletion: Adding or removing an item takes O(log n) time, where 'n' is the number of items. This is because an item only needs to travel up or down a logarithmic number of levels in the conceptual binary tree to restore the heap property.
