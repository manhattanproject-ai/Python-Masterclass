# OrderedDict

What is OrderedDict in Python?
OrderedDict is a dictionary subclass available in the collections module that remembers the order in which keys were first inserted. When you iterate through an OrderedDict, the items will be returned in the order they were added to the dictionary.


Key Characteristics of OrderedDict:

* Ordered: Maintains the insertion order of key-value pairs. This is the primary difference from a regular dict (before Python 3.7).
* Mutable: You can add, remove, and update key-value pairs.
* Unique Keys: Like regular dictionaries, keys must be unique.
* Heterogeneous Values: Values can be of any data type.
* Key Access: Values are accessed using their keys, just like regular dictionaries.
* Order-Dependent Equality: Equality testing between OrderedDict objects is order-sensitive. Two OrderedDict instances are considered equal only if they have the same key-value pairs and the pairs are in the same order. Equality testing with a regular dict is order-insensitive.

# 1. Define a Utility function to print highlighted characteristics

```python
import collections

def highlight_characteristic(name, description, symbol='='):
    """Helper function to print highlighted characteristics."""
    print(f"\n{symbol*5} {name.upper()} {symbol*5}")
    print(description)
    print(f"{symbol*5} {name.upper()} {symbol*5}\n")

print("--- Python OrderedDict Data Structure: Config File Loader Example ---")
```

# 2. Ordered (Insertion Order Guaranteed):

* config_settings is built by adding keys in a specific sequence.
* When print()ed or iterated over, the key-value pairs consistently appear in that exact insertion order.
* config_settings.move_to_end('db_host') is a method specific to OrderedDict (not available in standard dict) that demonstrates its ability to re-order elements while maintaining efficiency, highlighting its order-awareness.

```python

# 1. Ordered Characteristic

highlight_characteristic(
    "Ordered (Insertion Order Guaranteed)",
    "OrderedDicts remember the order in which key-value pairs were inserted. When iterated, elements are retrieved in this original order."
)

# Imagine loading config parameters line by line
config_settings = collections.OrderedDict()

print("Adding config settings in a specific order:")
config_settings['db_host'] = 'localhost'
config_settings['db_port'] = 5432
config_settings['api_key'] = 'abcdef12345'
config_settings['timeout_seconds'] = 30
config_settings['log_level'] = 'INFO'

print(f"Config Settings: {config_settings}")

print("\nIterating through config settings (order is preserved):")
for key, value in config_settings.items():
    print(f"  {key}: {value}")

# Demonstrate moving an item to end (specific to OrderedDict)
config_settings.move_to_end('db_host')
print(f"\nAfter moving 'db_host' to end: {config_settings}")
print("This operation maintains the relative order of other items while moving one item.")
```

# 3. Mutable (Changeable):

* Adding: config_settings['debug_mode'] = True adds a new key-value pair.
* Updating: config_settings['timeout_seconds'] = 60 modifies the value associated with an existing key. The key's position in the order does not change.
* Removing: del config_settings['api_key'] removes a pair by key, and config_settings.pop('db_port') removes a pair and returns its value.

```python
# 2. Mutable (Changeable) Characteristic

highlight_characteristic(
    "Mutable (Changeable)",
    "OrderedDicts can be modified after creation: new key-value pairs can be added, values can be updated, and pairs can be removed."
)

print(f"Current Config Settings: {config_settings}")

# Adding a new setting
config_settings['debug_mode'] = True
print(f"After adding 'debug_mode': {config_settings}")

# Updating an existing setting's value
config_settings['timeout_seconds'] = 60 # Increase timeout
print(f"After updating 'timeout_seconds': {config_settings}")

# Removing a setting
del config_settings['api_key'] # API key no longer needed directly in config
print(f"After removing 'api_key': {config_settings}")

# Using pop() to remove and get the value
removed_setting = config_settings.pop('db_port')
print(f"After removing 'db_port' using pop(): {config_settings}")
print(f"Removed port setting: {removed_setting}\n")
```

# 4. Unique Keys:

* When config_settings['log_level'] = 'DEBUG' is executed, it updates the value for the existing log_level key rather than creating a new entry. This reinforces that keys are unique.

```python
# 3. Unique Keys Characteristic

highlight_characteristic(
    "Unique Keys",
    "Each key in an OrderedDict must be unique. Adding a new value for an existing key will update the existing value."
)

print(f"Current Config Settings: {config_settings}")
config_settings['log_level'] = 'DEBUG' # Update existing key's value
print(f"After updating 'log_level' to 'DEBUG': {config_settings} (Note: no new entry, just value change)")

config_settings['debug_mode'] = False # Update existing key's value
print(f"After updating 'debug_mode' to 'False': {config_settings}")

# The order of the key remains the same when its value is updated.
# To change its order, you'd use move_to_end() explicitly.
```

# 5. Heterogeneous Values:

* The mixed_config OrderedDict demonstrates that values can be of any Python data type: str, int, bool, float, list, and dict.

```python
# 4. Heterogeneous Values Characteristic

highlight_characteristic(
    "Heterogeneous Values",
    "The values stored in an OrderedDict can be of different data types."
)

mixed_config = collections.OrderedDict()
mixed_config['service_name'] = "UserService"  # String
mixed_config['max_connections'] = 100         # Integer
mixed_config['is_enabled'] = True             # Boolean
mixed_config['version'] = 1.2                 # Float
mixed_config['allowed_ips'] = ['192.168.1.1', '10.0.0.5'] # List
mixed_config['credentials'] = {'user': 'admin', 'pass': 'secret'} # Dictionary

print(f"Mixed Config: {mixed_config}")
print(f"Type of 'max_connections': {type(mixed_config['max_connections'])}")
print(f"Type of 'allowed_ips': {type(mixed_config['allowed_ips'])}")
print(f"Type of 'credentials': {type(mixed_config['credentials'])}\n")
```

# 6. Key Access:

* Values are retrieved using their corresponding keys within square brackets (e.g., mixed_config['service_name']).
* The get() method is also shown for safer access, returning None or a specified default if the key is not found.

```python
# 5. Key Access Characteristic

highlight_characteristic(
    "Key Access",
    "Values are accessed using their unique keys, similar to a standard dictionary."
)

print(f"Accessing 'service_name': {mixed_config['service_name']}")
print(f"Accessing 'credentials' (and a sub-key): {mixed_config['credentials']['user']}")

# Using get() method for safe access
non_existent_key = mixed_config.get('non_existent_key', 'Default Value if not found')
print(f"Accessing 'non_existent_key' with default: {non_existent_key}")
```

# 7. Order-Dependent Equality:

* This is a key differentiator. od1 and od2 have the exact same key-value pairs but are False when compared using == because their insertion order differs.
* od1 and od3 are True because both their content and their insertion order match.
* This is contrasted with standard dicts (d1 and d2), where equality is solely based on content, regardless of insertion order (even in Python 3.7+ where they are insertion-ordered internally). This difference in __eq__ behavior is often the primary reason to still choose OrderedDict explicitly if order-based equality is a strict requirement.

```python
# 6. Order-Dependent Equality Characteristic (Crucial Distinction)

highlight_characteristic(
    "Order-Dependent Equality",
    "Two OrderedDicts are considered equal ONLY if they have the same key-value pairs AND the same insertion order. This differs from standard dictionaries in Python < 3.7 where order didn't matter for equality, and is still a subtle difference even in modern Python where 'dict' equality is purely content-based."
)

od1 = collections.OrderedDict()
od1['a'] = 1
od1['b'] = 2

od2 = collections.OrderedDict()
od2['b'] = 2
od2['a'] = 1 # Same content, different order of insertion

od3 = collections.OrderedDict()
od3['a'] = 1
od3['b'] = 2 # Same content, same order as od1

print(f"OrderedDict 1: {od1}")
print(f"OrderedDict 2: {od2}")
print(f"OrderedDict 3: {od3}")

print(f"od1 == od2? {od1 == od2} (False, due to order)")
print(f"od1 == od3? {od1 == od3} (True, same content and order)")

# Contrast with standard dict (in Python 3.7+, both will be insertion ordered, but their equality behavior is different for Ordered vs regular dict)
d1 = {'a': 1, 'b': 2}
d2 = {'b': 2, 'a': 1}
print(f"\nStandard dict 1: {d1}")
print(f"Standard dict 2: {d2}")
print(f"d1 == d2? {d1 == d2} (True, standard dict equality is content-based only)")
```
