# Stack

What is a Stack?

A Stack is a linear data structure that follows the Last-In, First-Out (LIFO) principle. This means that the most recently added element is the first one to be removed. Think of it like a stack of items where you can only access the top.   

Key Operations of a Stack:

* Push: Adding a new element to the top of the stack.
* Pop: Removing and returning the element from the top of the stack.
* Peek (or top): Viewing the element at the top of the stack without removing it.
* isEmpty: Checking if the stack is empty.
* size: Getting the number of elements in the stack.

# 1. Queueue Class and method

let's create a simple example of a Stack data structure in Python, using a common scenario: Browser History Navigation (like hitting the "back" button).

A. _history = [] (Internal Implementation):

We use a standard Python list to represent the stack. Python lists are very efficient for append() (add to end) and pop() (remove from end) operations, which perfectly map to push and pop in a stack. Both operations are O(1) (constant time) on average.

B. push(url):

This operation adds an element to the "top" of the stack.
It uses self._history.append(url). In our scenario, when you visit a new URL, it's added to the end of the list, becoming the new "top" of your history.

C. pop():

This operation removes and returns the element from the "top" of the stack.
It uses self._history.pop(). In our scenario, this simulates clicking the "back" button: the current page is removed from history, and you move to the previous one. It includes a check for an empty stack to prevent errors.

D. peek():

This operation allows you to look at (but not remove) the element at the "top" of the stack.
It accesses the last element using self._history[-1]. In our scenario, it tells you what page you are currently on without actually navigating away. It also handles the empty stack case.

E. is_empty():

This operation checks if the stack currently contains any elements.
It does this by checking if len(self._history) == 0.

F. size():

This operation returns the total number of elements currently in the stack.
It uses len(self._history)

```python
def print_separator(text):
    print(f"\n--- {text.upper()} ---")

print("--- Stack Data Structure: Browser History Navigation Example ---")

class BrowserHistoryStack:
    """
    Implements a Stack data structure using a Python list.
    Scenario: Tracking web pages visited to support a 'back' button.
    """
    def __init__(self):
        # A Python list naturally supports stack operations efficiently:
        # append() for push (add to end/top)
        # pop() for pop (remove from end/top)
        self._history = []
        print("Browser History Stack created (empty).")

    def push(self, url: str):
        """
        Operation: Push
        Adds a new page (element) to the top of the stack (the current page).
        """
        self._history.append(url)
        print(f"PUSH: Visited '{url}' (added to history).")
        self._display_current_state()

    def pop(self) -> str | None:
        """
        Operation: Pop
        Removes and returns the page (element) from the top of the stack (the current page),
        simulating going 'back' to the previous page.
        Returns None if the stack is empty.
        """
        if self.is_empty():
            print("POP: History is empty, cannot go back further.")
            return None
        page = self._history.pop() # pop() removes from the end
        print(f"POP: Going back from '{page}' (removed from history).")
        self._display_current_state()
        return page

    def peek(self) -> str | None:
        """
        Operation: Peek (or Top)
        Returns the page (element) at the top of the stack WITHOUT removing it.
        This is the current page you are on.
        Returns None if the stack is empty.
        """
        if self.is_empty():
            print("PEEK: History is empty, no current page.")
            return None
        current_page = self._history[-1] # Access the last element
        print(f"PEEK: Current page is '{current_page}'.")
        return current_page

    def is_empty(self) -> bool:
        """
        Operation: isEmpty
        Checks if the stack contains any elements.
        Returns True if the stack is empty, False otherwise.
        """
        is_empty = len(self._history) == 0
        print(f"isEmpty: History is {'empty' if is_empty else 'not empty'}.")
        return is_empty

    def size(self) -> int:
        """
        Operation: size
        Returns the number of pages (elements) currently in the history stack.
        """
        current_size = len(self._history)
        print(f"size: There are {current_size} pages in history.")
        return current_size

    def _display_current_state(self):
        """Helper to show the current state of the history stack."""
        # We display the stack from bottom to top (oldest to newest page)
        print(f"   Current History: {self._history} (Bottom -> Top/Current Page)\n")
```

# 2. Execute the methods

```python
# --- Demonstrate Stack Operations ---

# Create a new browser history stack
browser_history = BrowserHistoryStack()
browser_history.is_empty()
browser_history.size()

print_separator("Visiting Pages")
# Visit some pages (Push operations)
browser_history.push("google.com")
browser_history.push("youtube.com")
browser_history.push("youtube.com/watch?v=video1")

print_separator("Checking Current State")
# Check the current page and history size
browser_history.peek()
browser_history.size()

print_separator("Going Back")
# Go back a page (Pop operation)
last_page = browser_history.pop()
print(f"Confirmed back to: {last_page}")

# Go back another page
last_page = browser_history.pop()
print(f"Confirmed back to: {last_page}")

print_separator("Checking State After Going Back")
# Check current state again
browser_history.peek()
browser_history.size()
browser_history.is_empty()

print_separator("Trying to Go Back Too Far")
# Try to go back when the stack is empty
last_page = browser_history.pop() # Should return google.com
print(f"Confirmed back to: {last_page}")

browser_history.pop() # Stack should now be empty

print_separator("Final State")
browser_history.is_empty()
browser_history.size()
browser_history.peek()
```
