# Counter

What is Counter in Python?

* Counter is a dictionary subclass available in the collections module that is specifically designed for counting hashable objects. It takes an iterable (like a list or a string) as input and returns a dictionary where the keys are the unique elements from the iterable and the values are their counts.
* Think of it as a specialized dictionary that makes counting items really easy and provides some convenient methods for working with counts.

Key Characteristics of Counter:

* Dictionary Subclass: It inherits all the functionalities of a regular Python dictionary.
* Counts Hashable Objects: It can count the occurrences of any hashable object (like strings, numbers, tuples).
* Automatic Counting: When you initialize a Counter with an iterable, it automatically counts the occurrences of each unique item.
* Easy Updates: You can easily update the counts by adding counts from another iterable or another Counter.
* Provides Useful Methods: It offers methods like most_common() to find the most frequent elements and elements() to get an iterator over the elements repeated according to their counts.

# 1. Define a Utility function to print highlighted characteristics

```python
import collections

def highlight_characteristic(name, description, symbol='='):
    """Helper function to print highlighted characteristics."""
    print(f"\n{symbol*5} {name.upper()} {symbol*5}")
    print(description)
    print(f"{symbol*5} {name.upper()} {symbol*5}\n")

print("--- Python Counter Data Structure: Customer Feedback Analysis Example ---")
```

```python
# --- Scenario Setup: Raw Customer Feedback ---

customer_reviews = [
    "The product is excellent and easy to use. Excellent!",
    "Customer service was good, but the shipping was slow.",
    "Excellent quality and good value for money. Very good.",
    "The product met expectations. Easy to assemble."
]

# Simple tokenization (splitting into words and converting to lowercase for consistency)
# In a real scenario, you'd use more advanced NLP techniques (e.g., removing stopwords, stemming)
all_words = []
for review in customer_reviews:
    words = review.lower().replace('.', '').replace(',', '').split()
    all_words.extend(words)

print(f"Raw words from customer reviews: {all_words}\n")
```

# 2. Dictionary Subclass:

* word_counts is created and can be treated like a standard dictionary: you can access counts using square brackets (e.g., word_counts['excellent']), get lists of keys(), values(), and items().
* A key feature inherited (or enhanced) is that accessing a missing key returns 0 instead of raising a KeyError, which is extremely convenient for counting.

```python
# 1. Dictionary Subclass Characteristic

highlight_characteristic(
    "Dictionary Subclass",
    "Counter inherits from dict, so it behaves largely like a dictionary. It stores items as keys and their counts as values."
)

# Create a Counter instance
word_counts = collections.Counter(all_words)
print(f"Word Counts (Counter behaves like a dict): {word_counts}")

print(f"Accessing count for 'excellent': {word_counts['excellent']}") # Access like dict
print(f"Accessing count for 'good': {word_counts['good']}")
print(f"Accessing count for 'nonexistent_word': {word_counts['nonexistent_word']} "
      f"(Returns 0 for missing items, no KeyError)") # Important dict subclass feature

print(f"All words (keys): {list(word_counts.keys())}")
print(f"All counts (values): {list(word_counts.values())}")
print(f"All items: {list(word_counts.items())}\n")
```

# 3. Counts Hashable Objects:

* The example successfully counts str (words), int (sentiment scores), and tuple (user actions), demonstrating that Counter works with these hashable types. A comment notes that unhashable types like list cannot be counted directly.

```python
# 2. Counts Hashable Objects Characteristic

highlight_characteristic(
    "Counts Hashable Objects",
    "Counter is designed to count hashable items. These include strings, numbers, and tuples."
)

# The 'all_words' list contains strings, which are hashable.
# Example with numbers:
sentiment_scores = collections.Counter([1, 0, -1, 1, 1, 0, -1, 0, 1])
print(f"Sentiment Score Counts: {sentiment_scores}")

# Example with tuples (hashable):
user_actions = collections.Counter([('login', 'success'), ('logout', 'fail'), ('login', 'success')])
print(f"User Action Counts: {user_actions}\n")

# Attempting to count unhashable objects (e.g., lists) would lead to TypeError
# try:
#     invalid_counter = collections.Counter([['item1'], ['item2'], ['item1']])
# except TypeError as e:
#     print(f"Error caught for unhashable item: {e}")
```

# 4. Automatic Counting:

* When word_counts = collections.Counter(all_words) is executed, the Counter automatically iterates through all_words and tallies each unique word's occurrences without any manual loops or if/else checks.
* It also shows initialization from a mapping and keyword arguments.

```python
# 3. Automatic Counting Characteristic

highlight_characteristic(
    "Automatic Counting",
    "When initialized with an iterable, Counter automatically counts the occurrences of each item."
)

# This was shown in the first step:
# word_counts = collections.Counter(all_words)
print(f"Automatically counted words from 'all_words': {word_counts}\n")

# You can also initialize with a mapping or keyword arguments
another_counter = collections.Counter({'apple': 3, 'banana': 1})
print(f"Initialized from mapping: {another_counter}")

keyword_counter = collections.Counter(red=4, blue=2)
print(f"Initialized with keyword args: {keyword_counter}\n")
```

# 5. Easy Updates:

* word_counts.update(new_words_processed) efficiently adds the counts from new data to the existing counts.
* word_counts.subtract(issues_processed) decrements counts. This is particularly useful for tracking inventory or deficits, as counts can even go negative.

```python
# 4. Easy Updates Characteristic

highlight_characteristic(
    "Easy Updates",
    "Counters can be easily updated with new data using update(), or counts can be decremented using subtract()."
)

# Simulate receiving more feedback
new_feedback_words = ["Excellent", "excellent", "shipping", "fast", "service"]
new_words_processed = [word.lower() for word in new_feedback_words]

print(f"Word counts BEFORE update: {word_counts}")
word_counts.update(new_words_processed) # Add new counts
print(f"Word counts AFTER update with new feedback: {word_counts}")
print(f"Count of 'excellent' increased, 'shipping' count also changed.\n")

# Simulate a batch of negative reviews/issues (decreasing counts)
issues_detected = ["slow", "bug", "bug", "service"]
issues_processed = [issue.lower() for issue in issues_detected]

print(f"Word counts BEFORE subtract: {word_counts}")
word_counts.subtract(issues_processed) # Subtract counts
print(f"Word counts AFTER subtracting issues: {word_counts}")
print(f"Note: Counts can go to zero or become negative, representing debts or deficits.\n")
```

# 6. Provides Useful Methods:

* most_common(n): Quickly identifies the top n most frequent items.
* elements(): Provides an iterator that yields each element as many times as its count, effectively "unrolling" the counter.
* Arithmetic Operations: The example demonstrates how Counter objects can be combined and compared using standard arithmetic operators (+, -, |, &), which perform set-like and summation-like operations on their counts. This is powerful for data analysis tasks like combining different datasets or finding net differences.

```python
# 5. Provides Useful Methods Characteristic

highlight_characteristic(
    "Provides Useful Methods",
    "Counter offers specialized methods for common counting tasks, such as finding most common items, iterating through elements, and arithmetic operations."
)

print(f"Final word counts: {word_counts}")

# most_common(n): Returns a list of the n most common elements and their counts.
print(f"\n3 Most Common Words: {word_counts.most_common(3)}")

# elements(): Returns an iterator over elements repeating each as many times as its count.
print("\nAll words (repeating by count) from elements():")
print(list(word_counts.elements())) # Convert to list for clear output

# Arithmetic operations: Combining and comparing Counters
positive_words = collections.Counter({'excellent': 5, 'good': 3, 'value': 2, 'easy': 2})
negative_words = collections.Counter({'slow': 2, 'poor': 1, 'bug': 2, 'service': 1})
neutral_words = collections.Counter({'the': 10, 'product': 5, 'was': 3})

print(f"\nPositive Words Counter: {positive_words}")
print(f"Negative Words Counter: {negative_words}")
print(f"Neutral Words Counter: {neutral_words}")

# Addition (+): Combines counts (sum of positive counts)
total_sentiment_words = positive_words + negative_words
print(f"Positive + Negative (sum of counts, positive only): {total_sentiment_words}")

# Subtraction (-): Subtracts counts (only positive results)
net_sentiment = positive_words - negative_words
print(f"Positive - Negative (counts where positive > negative): {net_sentiment}")

# Intersection (&): Min of corresponding counts (common elements, lowest count)
common_between_pos_neg = positive_words & negative_words
print(f"Positive & Negative (min of counts for common elements): {common_between_pos_neg}")

# Union (|): Max of corresponding counts (all elements, highest count)
all_unique_sentiment_words = positive_words | negative_words | neutral_words
print(f"Positive | Negative | Neutral (max of counts for all elements): {all_unique_sentiment_words}")
```
