# Queue

What is a Queue?

A Queue is a linear data structure that follows the First-In, First-Out (FIFO) principle. This means that the element that was added first to the queue will be the first one to be removed. Think of it like a real-world queue or line.

Key Operations of a Queue:

* Enqueue (or offer): Adding a new element to the rear (end) of the queue.
* Dequeue (or poll): Removing and returning the element from the front (beginning) of the queue.
* Peek (or front): Viewing the element at the front of the queue without removing it.
* isEmpty: Checking if the queue is empty.   
* size: Getting the number of elements in the queue.

# 1. Queueue Class and method

let's create a simple example of a Queue data structure in Python, using a common scenario: a Customer Service Call Center Queue.

A. enqueue(customer_name):

* This operation adds an element to the rear of the queue.
* It uses self._queue.append(customer_name). In our scenario, a new customer joins the end of the line.

B. dequeue():

* This operation removes and returns the element from the front of the queue.
* It uses self._queue.popleft(). In our scenario, the customer who has been waiting the longest (at the front) is now served and leaves the queue. If the queue is empty, it handles this gracefully by printing a message and returning None.

C. peek():

* This operation allows you to look at (but not remove) the element at the very front of the queue.
* It simply accesses self._queue[0]. In our scenario, it tells us who is next in line without them actually being served yet. It also handles the case of an empty queue.

D. is_empty():

* This operation checks if the queue currently contains any elements.
* It does this by checking if len(self._queue) == 0.

E. size():

This operation returns the total number of elements currently in the queue.
It uses len(self._queue)

```python
import collections

print("--- Queue Data Structure: Customer Service Call Center Example ---")

class CallCenterQueue:
    """
    Implements a Queue data structure using collections.deque for efficient operations.
    Scenario: Customers waiting for service at a call center.
    """
    def __init__(self):
        # collections.deque is an efficient way to implement a queue in Python
        # as it provides O(1) (constant time) for adding/removing from both ends.
        self._queue = collections.deque()
        print("Call Center Queue created (empty).")

    def enqueue(self, customer_name: str):
        """
        Operation: Enqueue
        Adds a customer (element) to the rear (end) of the queue.
        """
        self._queue.append(customer_name)
        print(f"ENQUEUE: Customer '{customer_name}' joined the queue.")
        self._display_current_queue_state()

    def dequeue(self) -> str | None:
        """
        Operation: Dequeue
        Removes and returns the customer (element) from the front (beginning) of the queue.
        Returns None if the queue is empty.
        """
        if self.is_empty():
            print("DEQUEUE: Queue is empty, no customers to serve.")
            return None
        customer = self._queue.popleft() # popleft() removes from the front
        print(f"DEQUEUE: Customer '{customer}' is now being served (removed from queue).")
        self._display_current_queue_state()
        return customer

    def peek(self) -> str | None:
        """
        Operation: Peek (or Front)
        Returns the customer (element) at the front of the queue WITHOUT removing it.
        Returns None if the queue is empty.
        """
        if self.is_empty():
            print("PEEK: Queue is empty, no one at the front.")
            return None
        customer = self._queue[0] # Access the first element
        print(f"PEEK: Customer '{customer}' is currently at the front of the queue.")
        return customer

    def is_empty(self) -> bool:
        """
        Operation: isEmpty
        Checks if the queue contains any elements.
        Returns True if the queue is empty, False otherwise.
        """
        is_empty = len(self._queue) == 0
        print(f"isEmpty: Queue is {'empty' if is_empty else 'not empty'}.")
        return is_empty

    def size(self) -> int:
        """
        Operation: size
        Returns the number of customers (elements) currently in the queue.
        """
        current_size = len(self._queue)
        print(f"size: There are {current_size} customers in the queue.")
        return current_size

    def _display_current_queue_state(self):
        """Helper to show the current state of the queue."""
        print(f"   Current Queue: {list(self._queue)} (Front -> Rear)\n")
```

# 2. Execute the methods

```python
# --- Demonstrate Queue Operations ---

# Create a new call center queue
call_queue = CallCenterQueue()
call_queue.is_empty()
call_queue.size()

# Customer 1 calls in
call_queue.enqueue("Alice")

# Customer 2 calls in
call_queue.enqueue("Bob")

# Check who's at the front and how many are waiting
call_queue.peek()
call_queue.size()

# Customer 3 calls in
call_queue.enqueue("Charlie")

# Serve the next customer
served_customer = call_queue.dequeue()
print(f"Confirmed served: {served_customer}")

# Serve another customer
served_customer = call_queue.dequeue()
print(f"Confirmed served: {served_customer}")

# Try to peek when the queue is about to be empty
call_queue.peek()
call_queue.size()
call_queue.is_empty()

# Serve the last customer
served_customer = call_queue.dequeue()
print(f"Confirmed served: {served_customer}")

# Try to dequeue from an empty queue
call_queue.dequeue()
call_queue.is_empty()
call_queue.size()
```
