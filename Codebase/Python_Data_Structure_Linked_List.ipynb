# 1. Initialization

Node Class (The Building Block):

* Each Node object represents a single item on our shopping list.
* self.data: Stores the actual item (e.g., "Milk", "Eggs").
* self.next: This is the crucial part. It's a reference (a pointer) to the next Node in the sequence. If it's the last node, self.next will be None.

ShoppingList Class (The Manager):

* self.head: This is the entry point to our entire list. It's a reference to the very first Node. If self.head is None, it means the list is empty.


Let's create a simple example of a Linked List data structure in Python. We'll use a scenario of managing a shopping list where you can add, remove, and view items.

```python
def print_section(title):
    print(f"\n--- {title.upper()} ---")

print("--- Linked List Data Structure: Shopping List Example ---")

# 1. Node Class: The basic building block of a Linked List
class Node:
    """
    Represents a single item in our shopping list.
    Each node stores data (the item name) and a reference (pointer)
    to the next item in the list.
    """
    def __init__(self, data):
        self.data = data
        self.next = None  # Initially, this node doesn't point to anything

# 2. LinkedList Class: Manages the sequence of nodes
class ShoppingList:
    """
    Manages the overall shopping list.
    It keeps track of the 'head' (first item) of the list.
    """
    def __init__(self):
        self.head = None # The list starts empty, so no first item yet

    def is_empty(self) -> bool:
        """Checks if the shopping list is empty."""
        empty = self.head is None
        print(f"Is the shopping list empty? {empty}")
        return empty

    def append(self, item: str):
        """
        Operation: Append (Add to End)
        Adds a new item to the very end of the shopping list.
        """
        new_node = Node(item)
        if self.head is None:
            self.head = new_node # If list is empty, new node becomes the head
            print(f"APPEND: '{item}' added as the first item.")
            return

        # Traverse to the end of the list
        current = self.head
        while current.next:
            current = current.next
        current.next = new_node # Link the last node to the new node
        print(f"APPEND: '{item}' added to the end of the list.")

    def prepend(self, item: str):
        """
        Operation: Prepend (Add to Beginning)
        Adds a new item to the very beginning (front) of the shopping list.
        """
        new_node = Node(item)
        new_node.next = self.head # New node points to the old head
        self.head = new_node      # New node becomes the new head
        print(f"PREPEND: '{item}' added to the beginning of the list.")

    def delete_item(self, item_to_delete: str):
        """
        Operation: Delete
        Removes the first occurrence of a specific item from the list.
        """
        current = self.head
        previous = None

        # Case 1: Item to delete is the head
        if current and current.data == item_to_delete:
            self.head = current.next # Move head to the next node
            print(f"DELETE: '{item_to_delete}' removed from the beginning.")
            return

        # Case 2: Item is somewhere in the middle or end
        while current and current.data != item_to_delete:
            previous = current
            current = current.next

        # If item was not found
        if current is None:
            print(f"DELETE: '{item_to_delete}' not found in the list.")
            return

        # If item found, link previous node to next node, skipping current
        previous.next = current.next
        print(f"DELETE: '{item_to_delete}' removed from the list.")

    def display(self):
        """
        Operation: Display (Traversal)
        Prints all items in the shopping list from start to end.
        """
        if self.head is None:
            print("DISPLAY: Shopping list is empty.")
            return

        items = []
        current = self.head
        while current:
            items.append(current.data)
            current = current.next
        print(f"DISPLAY: Your shopping list: {' -> '.join(items)}")

    def search(self, item_to_find: str) -> bool:
        """
        Operation: Search
        Checks if a specific item exists in the shopping list.
        """
        current = self.head
        position = 0
        while current:
            if current.data == item_to_find:
                print(f"SEARCH: '{item_to_find}' found in the list at position {position}.")
                return True
            current = current.next
            position += 1
        print(f"SEARCH: '{item_to_find}' not found in the list.")
        return False

    def count_items(self) -> int:
        """
        Operation: Size
        Returns the total number of items currently in the shopping list.
        """
        count = 0
        current = self.head
        while current:
            count += 1
            current = current.next
        print(f"COUNT: There are {count} items in the shopping list.")
        return count
```

# 2. Demonstrate Linked List Operations :

* is_empty(): Checks if self.head is None.
* append(item):
  * Creates a new_node for the item.
  * If the list is empty (head is None), the new_node simply becomes the head.
  * Otherwise, it traverses the list from the head until it finds the very last node (current.next is None).
  * It then sets that last node's next pointer to the new_node, effectively adding the item to the end.
* prepend(item):
  * Creates a new_node for the item.
  * Sets the new_node's next pointer to the current head.
  * Makes the new_node the new head of the list. This is a very efficient O(1) operation.
* delete_item(item_to_delete):
  * Finds the node containing the item_to_delete and its previous node.
  * If the item is the head, it simply updates head to point to the next node.
  * If the item is in the middle or end, it bypasses the current node by making previous.next point directly to current.next. This effectively removes current from the chain.
* display():
  * Starts from the head and follows the next pointers of each node.
  * Collects and prints the data of each node until it reaches a node whose next is None (the end of the list).
* search(item_to_find):
  * Starts from the head and iterates through each node.
  * Compares the data of each node with item_to_find. If a match is found, it returns True. If it reaches the end without finding it, returns False.
* count_items():
  * Traverses the list from head to tail, incrementing a counter for each node encountered.

```python
# --- Demonstrate Linked List Operations ---

# Create an empty shopping list
my_shopping_list = ShoppingList()
my_shopping_list.is_empty()
my_shopping_list.count_items()
my_shopping_list.display()

print_section("Adding Items")
# Append items to the list
my_shopping_list.append("Milk")
my_shopping_list.append("Eggs")
my_shopping_list.append("Bread")
my_shopping_list.display()

# Prepend an item
my_shopping_list.prepend("Coffee")
my_shopping_list.display()

print_section("Checking and Searching")
my_shopping_list.is_empty()
my_shopping_list.count_items()
my_shopping_list.search("Eggs")
my_shopping_list.search("Cheese")

print_section("Deleting Items")
# Delete an item from the middle
my_shopping_list.delete_item("Eggs")
my_shopping_list.display()

# Delete the first item
my_shopping_list.delete_item("Coffee")
my_shopping_list.display()

# Delete a non-existent item
my_shopping_list.delete_item("Juice")
my_shopping_list.display()

# Delete the last item
my_shopping_list.delete_item("Bread")
my_shopping_list.display()

# Delete the very last item (makes list empty)
my_shopping_list.delete_item("Milk")
my_shopping_list.display()
my_shopping_list.is_empty()
my_shopping_list.count_items()
```
