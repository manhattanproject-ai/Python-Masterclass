# Tuple

What is a Tuple in Python?

A Tuple in Python is an ordered, immutable sequence of elements. Think of it as a container for a fixed collection of items.

Key Characteristics of Python Tuples:

* Ordered: The items in a tuple have a specific order, and this order is maintained. Elements are accessed by index (0-based).
* Immutable (Unchangeable): Once a tuple is created, you cannot add, remove, or change its elements. This is the main distinction from lists.
* Allows Duplicates: Like lists, tuples can contain multiple items with the same value.
* Heterogeneous: A single tuple can contain items of different data types (e.g., numbers, strings, booleans).
* Fixed Size (in terms of elements): Because they are immutable, the number of elements in a tuple is fixed upon creation.

# 1. Define a Utility function to print highlighted characteristics

```python
import collections

def highlight_characteristic(name, description, symbol='='):
    """Helper function to print highlighted characteristics."""
    print(f"\n{symbol*5} {name.upper()} {symbol*5}")
    print(description)
    print(f"{symbol*5} {name.upper()} {symbol*5}\n")

print("--- Python Tuple Data Structure: Product Order Record Example ---")
```

# 2. Exploring the Ordered nature of Tuple

Ordered:

* Elements are defined in a specific sequence (1001, "Laptop XYZ", ...).
* Accessing elements by their index (order_record[0], order_record[1]) always retrieves them in the same order, confirming their position is preserved.

```python
# 1. Ordered Characteristic

highlight_characteristic(
    "Ordered",
    "Elements in a tuple maintain their insertion order. The position of an element is significant and fixed."
)

# Creating an order record tuple
# Format: (Order ID, Product Name, Quantity, Price Per Unit, Customer ID)
order_record = (1001, "Laptop XYZ", 2, 1200.50, "CUST789")
print(f"Original Order Record Tuple: {order_record}")

print("\nAccessing elements by index (order is preserved):")
print(f"Order ID (index 0): {order_record[0]}")
print(f"Product Name (index 1): {order_record[1]}")
print(f"Quantity (index 2): {order_record[2]}")
print(f"Price Per Unit (index 3): {order_record[3]}")

# Another example to show order:
order_record_2 = ("Desk Lamp", 1, 35.99, 1002, "CUST790") # Different order of types
print(f"Another Order Record Tuple: {order_record_2}")
print(f"Order ID (index 3 here): {order_record_2[3]}")
```

# 3. Exploring the mutable nature of Tuple

Immutable:

* Attempts to modify an element by index (order_record[1] = "..."), or use methods like append() or remove(), result in TypeError or AttributeError, explicitly demonstrating that tuples cannot be changed after creation.
* Crucial Caveat: The example with order_with_mutable shows that if a tuple contains a mutable object (like a list), that mutable object itself can be altered. However, the tuple's structure and the references it holds remain unchanged. The tuple is still immutable in the sense that you can't reassign its elements to point to something else or change its length.

```python
# 2. Immutable Characteristic

highlight_characteristic(
    "Immutable",
    "Once a tuple is created, its elements cannot be changed (modified, added, or removed)."
)

print(f"Attempting to modify an element in the order_record tuple: {order_record}")
try:
    order_record[1] = "Desktop PC ABC" # Trying to change product name
except TypeError as e:
    print(f"Error caught as expected: {e}")
    print("This confirms that tuples are immutable - elements cannot be changed.")

print("\nAttempting to add an element:")
try:
    order_record.append("New Field") # Tuples do not have an append method
except AttributeError as e:
    print(f"Error caught as expected: {e}")
    print("This confirms that tuples are immutable - elements cannot be added.")

print("\nAttempting to remove an element:")
try:
    order_record.remove("Laptop XYZ") # Tuples do not have a remove method
except AttributeError as e:
    print(f"Error caught as expected: {e}")
    print("This confirms that tuples are immutable - elements cannot be removed.")

print(f"The original tuple remains unchanged: {order_record}")

# Note: If a tuple contains mutable objects (like a list), the mutable object itself can be changed,
# but the reference to that object within the tuple cannot.
highlight_characteristic(
    "Immutable (with a caveat)",
    "If a tuple contains a mutable object (like a list), that mutable object can be changed, "
    "but the tuple's structure itself (its elements' identities/order) remains fixed.", symbol='-')

order_with_mutable = (2001, "Custom Bundle", ["ItemA", "ItemB"], 500.00)
print(f"Tuple with a mutable list: {order_with_mutable}")
try:
    order_with_mutable[2].append("ItemC") # Modifying the list INSIDE the tuple
    print(f"After modifying the list inside the tuple: {order_with_mutable}")
except TypeError as e:
    print(f"Error caught (should not happen for list append): {e}")
print("The list 'ItemA', 'ItemB' was modified, but the tuple itself wasn't re-assigned or structurally changed.")
```

# 4. Exploring Duplicates in Tuple

Allows Duplicates:

* The order_with_duplicates tuple clearly shows "Smartphone X" appearing twice and the integer 1 appearing twice, confirming that duplicate elements are permitted.
* The count() method further verifies this.

```python
# 3. Allows Duplicates Characteristic

highlight_characteristic(
    "Allows Duplicates",
    "A tuple can contain multiple occurrences of the same element."
)

# Example: An order with multiple items of the same product or duplicate IDs
order_with_duplicates = (3001, "Smartphone X", 1, 800.00, "CUST900", "Smartphone X", 1)
print(f"Order tuple with duplicate product name and quantity: {order_with_duplicates}")
print(f"Count of 'Smartphone X': {order_with_duplicates.count('Smartphone X')}")
print(f"Count of 1: {order_with_duplicates.count(1)}")
```

# 5. Exploring Heterogeneous Characteristics of Tuple

Heterogeneous:

* The order_record tuple contains int, str, float.
* The complex_order_data tuple expands on this, including bool, list, and dict elements, demonstrating that a tuple can hold elements of any valid Python data type.
* The type() function is used to inspect the data types of the elements.

```python
# 4. Heterogeneous Characteristic
highlight_characteristic(
    "Heterogeneous",
    "A tuple can store elements of different data types."
)

# Our order_record already demonstrates this:
# (Integer, String, Integer, Float, String)
print(f"Order record: {order_record}")
print(f"Type of Order ID (index 0): {type(order_record[0])}")
print(f"Type of Product Name (index 1): {type(order_record[1])}")
print(f"Type of Quantity (index 2): {type(order_record[2])}")
print(f"Type of Price Per Unit (index 3): {type(order_record[3])}")
print(f"Type of Customer ID (index 4): {type(order_record[4])}")

# Another example with more diverse types
complex_order_data = (
    "ORD456",            # String (Order ID)
    "Gaming PC",         # String (Product Name)
    1,                   # Integer (Quantity)
    True,                # Boolean (Is urgent?)
    999.99,              # Float (Discounted price)
    ["Keyboard", "Mouse"], # List (Accessories)
    {"location": "Warehouse A", "slot": "B12"} # Dictionary (Storage info)
)
print(f"Complex Order Data Tuple: {complex_order_data}")
print(f"Type of element at index 3: {type(complex_order_data[3])}")
print(f"Type of element at index 5: {type(complex_order_data[5])}")
print(f"Type of element at index 6: {type(complex_order_data[6])}")
```

# 6. Exploring the Dynamic size nature of Tuple

Fixed Size:

* The "Immutable" characteristic largely covers this. The core idea is that you cannot add or remove elements from a tuple in-place.
* Attempts to del order_record[0] (delete an element by index) result in a TypeError.
* The += operator on a tuple actually creates a new tuple by concatenating the original tuple with new elements, and then reassigns the variable to this new tuple. It does not modify the original tuple in memory. This is explicitly demonstrated by comparing the id() of the tuple before and after the += operation.

```python
# 5. Fixed Size Characteristic
highlight_characteristic(
    "Fixed Size",
    "The number of elements in a tuple cannot be changed after its creation. You cannot add or remove elements."
)

print(f"Original order_record tuple: {order_record}")
print(f"Size of order_record: {len(order_record)}")

# As demonstrated in 'Immutable' section, append/remove methods do not exist for tuples.
# Trying to add:
try:
    order_record += ("New Status",) # This creates a NEW tuple, doesn't modify the original
    print(f"Attempting to 'add' an element creates a NEW tuple: {order_record}")
    print(f"Size of new tuple: {len(order_record)}")
except TypeError as e:
    print(f"Error caught (unexpected here if += creates new): {e}") # This won't error

# To emphasize, `+=` with tuples creates a new tuple, it does not mutate the original in-place.
original_id = id(order_record)
print(f"Original ID: {original_id}")
order_record = order_record + ("Shipping Pending",) # This operation creates a new tuple and reassigns the variable
new_id = id(order_record)
print(f"After 'adding' an element, new tuple ID: {new_id}")
print(f"Original tuple ID is DIFFERENT from new tuple ID, confirming a new tuple was created.")
print(f"New tuple: {order_record} (Size: {len(order_record)})")


# Trying to remove:
try:
    del order_record[0] # Trying to delete an element by index
except TypeError as e:
    print(f"Error caught as expected: {e}")
    print("Cannot delete individual elements from a tuple (Fixed Size).")

print(f"The size of the tuple remains fixed after creation regarding in-place modification. To 'change' it, a new tuple must be created.")
```
