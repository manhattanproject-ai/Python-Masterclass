# Hash Table

A Hash Table (also known as a Hash Map or Dictionary in some languages, including Python's built-in dict) is a data structure that uses a special function called a hash function to map keys to specific locations (indices or "buckets") in an array (the "table"). This mapping allows for very efficient retrieval, insertion, and deletion of data based on the key.

Key Concepts of Hash Tables:

* Key: The identifier for the data you want to store and retrieve (e.g., a username, a product ID). Keys should ideally be unique.
* Value: The actual data associated with the key (e.g., a user's profile, product details).
* Hash Function: A function that takes a key as input and produces an integer (the "hash value" or "hash code"). This hash value is then used to determine the index in the underlying array where the value should be stored. A good hash function aims to distribute keys evenly across the table to minimize collisions.
* Table (Bucket Array): The underlying array where the key-value pairs are stored. Each position in the array is often called a "bucket."
* Collision: Occurs when the hash function produces the same hash value for two different keys. Efficient hash table implementations need strategies to handle collisions.

# 1. Initialization :  Creates an empty hash table.

```python
def print_section(title):
    print(f"\n--- {title.upper()} ---")

print("--- Hash Table (Dictionary) Example: Student Gradebook ---")

# 1. Creating a Hash Table (Python's Dictionary)
# Our hash table will map Student IDs (Keys) to Student Information (Values)
# Key: Student ID (e.g., 'S001', 'S002') - must be unique and hashable
# Value: A tuple or list containing (Student Name, Grade)
student_gradebook = {} # This is an empty dictionary, serving as our hash table

print("Empty Student Gradebook (Hash Table) created:")
print(student_gradebook)
```

# 2. Insertion (Adding Key-Value Pairs)

* Insertion (student_gradebook['S001'] = ("Alice", 95)): Adds a new key-value pair. If the key already exists, its value is updated.

```python
# 2. Insertion (Adding Key-Value Pairs)

print_section("Adding Student Records (Insertion)")
student_gradebook['S001'] = ("Alice", 95)
print(f"Added S001: {student_gradebook['S001']}")

student_gradebook['S002'] = ("Bob", 88)
print(f"Added S002: {student_gradebook['S002']}")

student_gradebook['S003'] = ("Charlie", 72)
print(f"Added S003: {student_gradebook['S003']}")

student_gradebook['S004'] = ("Diana", 91)
print(f"Added S004: {student_gradebook['S004']}")

print("\nCurrent Gradebook:")
print(student_gradebook)
```

# 3. Lookup (Retrieving Values by Key)

* Lookup (student_gradebook['S002']): Retrieves the value associated with a given key. This is incredibly fast.

```python
# 3. Lookup (Retrieving Values by Key)

print_section("Looking Up Student Grades (Retrieval)")
student_id_to_find = 'S002'
if student_id_to_find in student_gradebook:
    name, grade = student_gradebook[student_id_to_find]
    print(f"Grade for {name} (ID: {student_id_to_find}): {grade}")
else:
    print(f"Student with ID {student_id_to_find} not found.")

student_id_to_find = 'S005' # Non-existent ID
if student_id_to_find in student_gradebook:
    name, grade = student_gradebook[student_id_to_find]
    print(f"Grade for {name} (ID: {student_id_to_find}): {grade}")
else:
    print(f"Student with ID {student_id_to_find} not found.")
```

# 4. Updating Values

* Updating (student_gradebook['S001'] = ("Alice", 98)): Simply assigning a new value to an existing key updates its entry.

```python
# 4. Updating Values

print_section("Updating a Student's Grade")
print(f"Original grade for S001 (Alice): {student_gradebook['S001'][1]}")
student_gradebook['S001'] = ("Alice", 98) # Update the grade
print(f"Updated grade for S001 (Alice): {student_gradebook['S001'][1]}")
print("\nCurrent Gradebook:")
print(student_gradebook)
```

# 5. Deletion (Removing Key-Value Pairs)

* Deletion (del student_gradebook['S003']): Removes a key-value pair from the hash table.

```python
# 5. Deletion (Removing Key-Value Pairs)

print_section("Removing a Student Record (Deletion)")
student_id_to_remove = 'S003'
if student_id_to_remove in student_gradebook:
    del student_gradebook[student_id_to_remove]
    print(f"Removed student with ID: {student_id_to_remove}")
else:
    print(f"Student with ID {student_id_to_remove} not found for deletion.")

print("\nGradebook after deletion:")
print(student_gradebook)

# Verify deletion
print(f"Does S003 still exist? {'S003' in student_gradebook}") # Expected: False
```

# 6. Checking for Key Existence

* Checking Existence ('S001' in student_gradebook): Efficiently checks if a key is present in the hash table.

```python
# 6. Checking for Key Existence

print_section("Checking for Key Existence")
print(f"Is 'S001' in gradebook? {'S001' in student_gradebook}") # Expected: True
print(f"Is 'S003' in gradebook? {'S003' in student_gradebook}") # Expected: False
print(f"Is 'S004' in gradebook? {'S004' in student_gradebook}") # Expected: True
```

# 7. Iterating Through the Hash Table

Iteration (for student_id, info in student_gradebook.items():): While retrieval by key is its primary strength, you can also iterate through all key-value pairs stored in the hash table.

```python
# 7. Iterating Through the Hash Table
print_section("Iterating Through the Gradebook")
print("Student IDs and their info:")
for student_id, info in student_gradebook.items():
    name, grade = info
    print(f"ID: {student_id}, Name: {name}, Grade: {grade}")
```
