# Numpy array

What are NumPy Arrays?

* NumPy (short for Numerical Python) is a powerful library in Python that provides support for large, multi-dimensional arrays and matrices, along with a large collection of high-level mathematical functions to operate on these arrays efficiently. A NumPy array is the core data structure of the NumPy library.   
* Think of a NumPy array as a grid of values, all of the same type, organized in one or more dimensions. This homogeneity is a key difference from Python lists, which can hold items of different types. This uniformity allows NumPy to perform operations much faster.

Key Characteristics of NumPy Arrays:

* Homogeneous: All elements in a NumPy array must be of the same data type (e.g., all integers, all floating-point numbers). This makes operations more efficient.
* Multi-dimensional: NumPy arrays can have one dimension (like a list), two dimensions (like a table or a matrix), or even more dimensions (for representing higher-dimensional data).
* Fixed Size (at creation): While you can reshape arrays, the total number of elements is usually fixed when you create the array.
* Efficient Numerical Operations: NumPy is highly optimized for mathematical and logical operations on entire arrays. This is done using vectorized operations, which are much faster than iterating through Python lists.
* Memory Efficient: Due to the homogeneous nature and optimized storage, NumPy arrays can be more memory-efficient than Python lists for large numerical datasets.
* Foundation for Scientific Computing: NumPy arrays are the fundamental data structure used by many other scientific and data analysis libraries in Python, such as Pandas, SciPy, and scikit-learn.

# 1. Define a Utility function to print highlighted characteristics

```python
import numpy as np
import sys # For demonstrating memory efficiency (conceptually)

def highlight_characteristic(name, description, symbol='='):
    """Helper function to print highlighted characteristics."""
    print(f"\n{symbol*5} {name.upper()} {symbol*5}")
    print(description)
    print(f"{symbol*5} {name.upper()} {symbol*5}\n")

print("--- Python NumPy Arrays: Sensor Data Processing Example ---")
```

```python
# --- Scenario Setup: Simulating Sensor Readings ---
# Imagine 3 sensors taking 5 temperature readings each (e.g., hourly)
sensor_data_2d = np.array([
    [22.5, 23.1, 22.8, 23.0, 22.7],  # Sensor 1: Temperatures
    [21.9, 22.3, 22.0, 22.5, 22.1],  # Sensor 2: Temperatures
    [23.0, 23.5, 23.2, 23.8, 23.3]   # Sensor 3: Temperatures
])
print(f"Original 2D Sensor Data Array (Temperatures):\n{sensor_data_2d}\n")
```

# 2. Homogeneous:

* sensor_data_2d is initialized with float values, and its dtype is confirmed as float64.
* The mixed_data example shows that when an array is created with mixed types (int and float), NumPy automatically "upcasts" all elements to the common, most general type (float in this case).
* Similarly, if strings are introduced, all elements are converted to a string data type.

```python
# 1. Homogeneous Characteristic

highlight_characteristic(
    "Homogeneous",
    "All elements in a NumPy array must be of the same data type. If different types are provided, NumPy will automatically upcast to a common, compatible type (e.g., integer to float if floats are present)."
)

print(f"Data type (dtype) of sensor_data_2d: {sensor_data_2d.dtype}") # Will be float64 by default

# Example of type casting/upcasting
mixed_data = np.array([1, 2, 3.5, 4]) # Contains int and float
print(f"Mixed data array: {mixed_data}")
print(f"Data type of mixed_data (upcasted to float): {mixed_data.dtype}")

string_data = np.array([1, 2, 'hello', 4]) # Contains int and string
print(f"String data array (upcasted to string): {string_data}")
print(f"Data type of string_data: {string_data.dtype}\n")
```

# 3. Multi-dimensional:

* Examples of 1D (single_sensor_readings), 2D (sensor_data_2d), and 3D (environmental_data_3d) arrays are shown.
* The shape attribute explicitly shows the dimensions of the array (e.g., (3, 5) for a 3x5 matrix).
* The ndim attribute confirms the number of dimensions.

```python
# 2. Multi-dimensional Characteristic

highlight_characteristic(
    "Multi-dimensional",
    "NumPy arrays can represent data in one, two, three, or even higher dimensions. They are the fundamental data structure for vectors, matrices, and tensors."
)

# 1D Array (Vector) - e.g., a single sensor's readings
single_sensor_readings = np.array([22.5, 23.1, 22.8, 23.0, 22.7])
print(f"1D Array (Vector):\n{single_sensor_readings}")
print(f"  Shape: {single_sensor_readings.shape} (5 elements)")
print(f"  Dimensions (ndim): {single_sensor_readings.ndim}\n")

# 2D Array (Matrix) - our main sensor_data_2d
print(f"2D Array (Matrix - Sensor Data):\n{sensor_data_2d}")
print(f"  Shape: {sensor_data_2d.shape} (3 rows, 5 columns)")
print(f"  Dimensions (ndim): {sensor_data_2d.ndim}\n")

# 3D Array (Tensor) - e.g., temperature, humidity, pressure readings over time for sensors
environmental_data_3d = np.array([
    [ # Location 1
        [22.5, 60.1, 1012.5], # Sensor 1, Time 1 (Temp, Hum, Pres)
        [23.1, 62.0, 1011.8]  # Sensor 1, Time 2
    ],
    [ # Location 2
        [21.9, 58.5, 1013.0], # Sensor 2, Time 1
        [22.3, 59.8, 1012.7]  # Sensor 2, Time 2
    ]
])
print(f"3D Array (Tensor - Environmental Data):\n{environmental_data_3d}")
print(f"  Shape: {environmental_data_3d.shape} (2 locations, 2 times, 3 metrics)")
print(f"  Dimensions (ndim): {environmental_data_3d.ndim}\n")
```

# 4. Fixed Size (at creation):

* The np.append() function is demonstrated. It does not modify the original sensor_data_2d array in place; instead, it returns a new array with the appended data.
* Similarly, reshape() returns a new view or copy, leaving the original array's dimensions untouched. This highlights that arrays' sizes are fixed once memory is allocated for them.

```python
# 3. Fixed Size (at creation) Characteristic

highlight_characteristic(
    "Fixed Size (at creation)",
    "Once a NumPy array is created, its size (number of elements and dimensions) is fixed. Operations that seem to change size (like reshaping) actually return a new view or a new array without modifying the original data's memory footprint or creating new data."
)

print(f"Original sensor_data_2d shape: {sensor_data_2d.shape}")
print(f"Original sensor_data_2d data:\n{sensor_data_2d}")

# Attempt to append (this creates a *new* array)
new_reading = np.array([24.0, 24.1, 24.2, 24.3, 24.4])
# np.append returns a new array, doesn't modify in place
appended_array = np.append(sensor_data_2d, [new_reading], axis=0)
print(f"\nResult of np.append (new array created):\n{appended_array}")
print(f"Original sensor_data_2d remains unchanged:\n{sensor_data_2d}")

# Attempt to resize (also creates a new array if elements are added/removed)
# If new size is smaller, it truncates; if larger, it fills with zeros/garbage.
# reshape() creates a view without copying data if possible.
reshaped_array = sensor_data_2d.reshape(5, 3) # Creates a new view if possible
print(f"\nReshaped sensor_data_2d (new view/array):\n{reshaped_array}")
print(f"Original sensor_data_2d remains unchanged:\n{sensor_data_2d}\n")
```

# 5. Memory efficient

When you run this updated code, especially for num_elements = 100000 (or more), you'll see a dramatic difference:

* The Python list will consume significantly more memory because of the overhead of each individual Python object.
* The NumPy array will remain very compact, directly proportional to number_of_elements * size_of_each_element_in_bytes.

```python
# 4. Memory Efficient Characteristic

num_elements = 100000 # Let's use 100,000 elements

# Python List
list_of_numbers_large = list(range(num_elements))
# To get a more accurate total memory for a list, sum the size of the list object and all its elements
total_list_mem = sys.getsizeof(list_of_numbers_large) + sum(sys.getsizeof(i) for i in list_of_numbers_large)
print(f"Total memory size of a Python list of {num_elements} integers: {total_list_mem} bytes")

# NumPy Array
# Let's explicitly define dtype for clarity, e.g., 32-bit integers
numpy_array_of_numbers_large = np.arange(num_elements, dtype=np.int32)
print(f"Memory size of a NumPy array of {num_elements} 32-bit integers: {sys.getsizeof(numpy_array_of_numbers_large)} bytes")

# Also compare with float64, which is common in scientific computing
numpy_array_of_floats_large = np.arange(num_elements, dtype=np.float64)
print(f"Memory size of a NumPy array of {num_elements} 64-bit floats: {sys.getsizeof(numpy_array_of_floats_large)} bytes")

# For comparison, a Python list of 100,000 floats
list_of_floats_large = [float(i) for i in range(num_elements)]
total_list_floats_mem = sys.getsizeof(list_of_floats_large) + sum(sys.getsizeof(i) for i in list_of_floats_large)
print(f"Total memory size of a Python list of {num_elements} floats: {total_list_floats_mem} bytes")
```

# 6. Foundation for Scientific Computing:

* Element-wise operations (ufuncs): Multiplying sensor_data_2d * 1.8 + 32 applies the conversion to every element in the array efficiently, without explicit loops.
* Aggregation functions: np.mean(), np.max(), np.sum() are used with axis parameters to perform calculations across specific dimensions (e.g., average per sensor or per hour), which is common in data analysis.
* Broadcasting: Adding a single offset value to the entire sensor_data_2d array demonstrates broadcasting, where NumPy intelligently expands the smaller array to match the shape of the larger one for the operation.
* Boolean Indexing: Comparing the array to a high_temp_threshold creates a boolean array, which can then be used for filtering or counting specific conditions (np.sum(is_high_temp)).
* Matrix Multiplication (np.dot / @): A fundamental operation in linear algebra, showing how NumPy handles complex mathematical computations efficiently.

```python
# 5. Foundation for Scientific Computing Characteristic

highlight_characteristic(
    "Foundation for Scientific Computing",
    "NumPy provides highly optimized functions for numerical operations, linear algebra, Fourier transforms, random number generation, etc., making it the backbone of scientific computing in Python."
)

print(f"Original Sensor Data (Temperatures):\n{sensor_data_2d}\n")

# Element-wise operations (Universal Functions - ufuncs)
# Convert Celsius to Fahrenheit: F = C * 1.8 + 32
fahrenheit_data = sensor_data_2d * 1.8 + 32
print(f"Temperatures in Fahrenheit:\n{fahrenheit_data}\n")

# Aggregation functions (e.g., mean, sum, max, min)
average_temp_per_sensor = np.mean(sensor_data_2d, axis=1) # Average across rows (per sensor)
print(f"Average temperature per sensor: {average_temp_per_sensor}")

average_temp_per_hour = np.mean(sensor_data_2d, axis=0) # Average across columns (per hour)
print(f"Average temperature per hour: {average_temp_per_hour}")

max_temp_overall = np.max(sensor_data_2d)
print(f"Maximum temperature recorded: {max_temp_overall}\n")

# Broadcasting (performing operations on arrays of different shapes)
# Add a fixed offset to all readings
offset = 0.5
adjusted_data = sensor_data_2d + offset
print(f"Sensor data + {offset} offset:\n{adjusted_data}\n")

# Compare data against a threshold
high_temp_threshold = 23.0
is_high_temp = sensor_data_2d > high_temp_threshold
print(f"Readings above {high_temp_threshold}C (Boolean array):\n{is_high_temp}")
print(f"Number of high temperature readings: {np.sum(is_high_temp)}\n")

# Matrix multiplication (Dot product) - fundamental for many scientific applications
matrix_A = np.array([[1, 2], [3, 4]])
matrix_B = np.array([[5, 6], [7, 8]])
matrix_product = np.dot(matrix_A, matrix_B) # or matrix_A @ matrix_B
print(f"Matrix A:\n{matrix_A}")
print(f"Matrix B:\n{matrix_B}")
print(f"Matrix Product (A @ B):\n{matrix_product}\n")
```
